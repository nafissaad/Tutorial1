```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df.rename(columns = {'per capita income (US$)':'GDP_Per_Cap'}, inplace = True)
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.plot(np.array(df.year),df[['GDP_Per_Cap']],color='red',marker='+')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']].values,df.GDP_Per_Cap.values)
```

```python
reg.predict([[2016]])
```
