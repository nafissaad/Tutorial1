```python
#Description: retire within 10 years
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#Get your savings rate
savings_rate=0.64
#get your yearly income
income=30000
#Get your yearly expenses
expenses=income*(1-savings_rate)
#Get the interest rate from ypur portfolio or assets
interest_rate=0.08
#get the amount that you will invest into the portfolio or asstes
investment=income-expenses
#calculate the expected annual returns
annual_return=investment*interest_rate
#Get the Current year
year=2022
#Get the withdrawl rate
withdrawl_rate=0.04
```

```python
#create empty list for the dataset
Year=[]
Yearly_Income=[]
Yearly_Expenses=[]
Yearly_Investment_Amount=[]
Total_Invested_Amount=[]
Annual_Returns=[]
```

```python
#Append the date to the list
Year.append(year)
Yearly_Income.append(income)
Yearly_Expenses.append(expenses)
Yearly_Investment_Amount.append(income - expenses)
Total_Invested_Amount.append(investment)
Annual_Returns.append(annual_return)
```

```python
#Create a variable to  store the amount of years/rows invested
invested_years=30
for i in range(0,invested_years-1):
    investment=(income-expenses)+annual_return+investment
    annual_return=investment*interest_rate
    #Appned the new data to the lists
    Year.append(year+i+1)
    Yearly_Income.append(income)
    Yearly_Expenses.append(expenses)
    Yearly_Investment_Amount.append(income - expenses)
    Total_Invested_Amount.append(investment)
    Annual_Returns.append(annual_return)
```

```python
#Create the set
df=pd.DataFrame()
df['Year']=Year
df['Yearly_income']=Yearly_Income
df['Yearly_Expenses']=Yearly_Expenses
df['Yearly_invested_Amount']=Total_Invested_Amount
df['Annual_Returns']=Annual_Returns
Yearly_Withdrawl_Amount=np.array(Total_Invested_Amount)*withdrawl_rate
df['Yearly_Withdrawal_Amount']=Yearly_Withdrawl_Amount
```

```python
#Visually Show the dat
plt.figure(figsize=(12,4))
plt.plot(np.array(df['Year']),df['Yearly_Expenses'].values,label='Yearly Expenses')
plt.plot(np.array(df['Year']),df['Yearly_Withdrawal_Amount'].values,label='Yearly_Withdrawl_Amount')
plt.title('Years Untill Retirement')
plt.xlabel('USD($)')
plt.xticks(df['Year'],rotation=45)
plt.legend()
plt.show()
```
