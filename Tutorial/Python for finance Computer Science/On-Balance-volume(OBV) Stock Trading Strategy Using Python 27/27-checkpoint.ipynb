```python
#This program uses on-Balance volume (OBV tp determine when to buy and sell stock
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
plt.style.use('fivethirtyeight')
```

```python
#Load the data
df= yf.Ticker('AMZN').history(start='2020-03-16',end='2020-09-19')
#df.set_index(pd.DatetimeIndex(df['Date'].values))
df
```

```python
#Visually show the stock Price 
plt.figure(figsize=(12.2,4.5))
plt.plot(df['Close'],label='Close')
plt.title('close Price')
plt.xlabel('Date',fontsize=18)
plt.ylabel('Price Use')
plt.show()
```

```python
#Calculate the on Balance Volume(OBV)
OBV=[]
OBV.append(0)
#loop through the data set (close price) from the second row(index 1) to the end of the data set
for i in range(1, len(df.Close)):
    if df.Close[i] > df.Close[i-1]:
        OBV.append(OBV[-1] + df.Volume[i])
    elif df.Close[i] < df.Close[i-1]:
        OBV.append(OBV[-1] - df.Volume[i])
    else:
        OBV.append(OBV[-1])
```

```python
#Store the OBV and OBV Exponential Moving Average (EMA into new columns)
df['OBV']=OBV
df['OBV_EMA']=df['OBV'].ewm(span=20).mean()
df
```

```python
#Visually show the stock Price 
plt.figure(figsize=(12.2,4.5))
plt.plot(df['OBV'],label='OBV',color='orange')
plt.plot(df['OBV_EMA'],label='OBV_EMA',color='purple')
plt.title('OBV/OBV EMA Chart')
plt.xlabel('Date',fontsize=18)
plt.ylabel('Price Use')
plt.show()
```

```python
#Create a function to signal when tobuy and sell the stock
#If OBV > OBV_EMA then BUY 
# if OBV < OBV_EMA Then Sell
#Else do Nothing 
def Buy_sell(signal,col1,col2):
    buy=[]
    sell=[]
    index=[]
    flag=-1
    for i in range(len(signal)):
        if signal[col1][i] > df[col2][i] and flag != 1:
            buy.append(signal['Close'][i])
            flag=1
        if signal[col1][i] < signal[col2][i] and flag != 0:
            sell.append(signal['Close'][i])
            flag=0
            index.append(i)
        else:
            buy.append(np.nan)
            sell.append(np.nan)
            
        #print(len(buy),signal['Close'][i],i)
        #print(len(sell),signal['Close'][1],i)
    return(buy,sell)

```

```python
#create Buy and sell columns
x = Buy_sell(df,'OBV','OBV_EMA')
df['Buy']=x[0]
df['Sell']=x[1]
#Show the data set
#Visually show the stock Price 
plt.figure(figsize=(12.2,4.5))
plt.plot(df['Close'],label='Close',alpha=0.35)
plt.scatter(df.index,df['Buy'],label='Buy',marker='^',alpha=1,color='green')
plt.scatter(df.index,df['Sell'],label='Sell',marker='v',alpha=1,color='red')
plt.title('OBV/OBV EMA Chart')
plt.xlabel('Date',fontsize=18)
plt.ylabel('Price USD',fontsize=18)
plt.show()
```
