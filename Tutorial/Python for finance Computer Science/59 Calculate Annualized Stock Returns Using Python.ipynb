```python
plt.plot(df1['Close'], label = "2018")
plt.plot(df2['Close'], label = "2019")
plt.plot(df3['Close'], label = "2020")
plt.plot(df4['Close'], label = "2021")
plt.legend()
```

```python
#Description: Calculate annualized returns
#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_historical_Data(tickers):
    """This function returns a pd dataframe with all of the adjusted closing information"""
    data = pd.DataFrame()
    names = list()
    for i in tickers:
        data = pd.concat([data, pd.DataFrame(yf.download(i, start=datetime(2018, 4, 24), end=datetime(2022, 12, 21)).iloc[:,4])], axis = 1)
        names.append(i)
    data.columns = names
    return data

ticks = ["Meta", "AMZN", "TSLA", "GOOG", "AAPL", "CRM", "NFLX", "MSFT", "NVDA","SPY","VGT","VOO","NKE", "WMT","ko", "MSFT"] #Name of company (Dominos pizza)
d = get_historical_Data(ticks)
print(d.shape)
# Most Recent Data
d.tail()
```

```python
d
```

```python
#Get the simple Daily returns of the assets
daily_simple_returns=d.pct_change()
#Show the daily simple returns
daily_simple_returns
```

```python
#Calculate The annualized returns
annualized_returns=daily_simple_returns.mean()*252
#Show the returns
annualized_returns
```

```python
#Sort the Data
sorted_annualized_returns=annualized_returns.sort_values(ascending=False)
sorted_annualized_returns
```

```python
#Plot the sorted annualized return
plt.bar(sorted_annualized_returns.index,sorted_annualized_returns)
plt.ylabel('Annualized Returns')
plt.xlabel('Assets')
plt.title('Annualized Returns pf Assets')
plt.xticks(rotation=90)
```
