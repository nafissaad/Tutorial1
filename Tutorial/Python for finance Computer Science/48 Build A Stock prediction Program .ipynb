```python
# this program predicts stock prices by  using machine learning models
# Install the dependencies
import yfinance as yf
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
#get the stock data
#load the data
df= yf.Ticker('META').history(start='2019-04-19',end='2022-10-14')
df.reset_index(inplace=True)
df
```

```python
#Get the Adjusted Close Price
df=df[['Close']]
#Take a look at the new data
print(df.head())
```

```python
#A variable for predicting
forecast_out= 30
#Create another column(the target or dependent variable) shift 'n' units up
df['Prediction']=df[['Close']].shift(-forecast_out)
#print the new data
print(df)
```

```python
#Create the indeoendent data set (x)
#Convert the dataframe to a numpy array
X=np.array(df.drop(['Prediction'],axis=1))
#Remove the last 'n' rows
X=X[:-forecast_out]
X
```

```python
# Create the depencent data set (y)###
#Convert the dataframe to a numpy array(all of the values including the NaN's)
y=np.array(df['Prediction'])
# Get all of the y values except the last 'n' rows
y=y[:-forecast_out]
y
```

```python
#Split the data into 80% training and 20% testing
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#Create and train the Support Vector Machine (Regressor)
svr_rbf=SVR(kernel='rbf',C=1e3,gamma=0.1)
svr_rbf.fit(x_train,y_train)
```

```python
svr_rbf.score(x_test,y_test)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
x_forecast=np.array(df.drop(['Prediction'],axis=1))[-forecast_out:]
x_forecast
```
