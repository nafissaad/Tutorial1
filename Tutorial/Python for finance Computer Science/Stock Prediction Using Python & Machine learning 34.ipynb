```python
#Import the liberires
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
df= yf.Ticker('AMZN').history(start='2010-01-01',end='2022-08-02')
df
```

```python
#Manipulate the data
#Create the target column
df['Price_up']=np.where(df['Close'].shift(-1)>df['Close'],1,0)
#show the data
df
```

```python
X=df.iloc[:,0:df.shape[1]-1].values
Y=df.iloc[:,df.shape[1]-1].values
X
```

```python
#Split the data again but this time into 80% training and 20% testing data sets
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
tree=DecisionTreeClassifier().fit(X_train,Y_train)
#show how well the model diid on the test data set
print(tree.score(X_test,Y_test))
```

```python
tee_predictions=tree.predict(X_test)
print(tee_predictions)
```
