```python
import pandas as pd
```

```python
#https://data.worldbank.org/indicator/NY.GDP.DEFL.KD.ZG?locations=BD
inflation_data=pd.read_csv('C:\\Users\\nafissaad\\Time Series\\API_NY.GDP.DEFL.KD.ZG_DS2_en_csv_v2_5551676.csv',header=2)
inflation_data.info()
```

```python
inflation_data[inflation_data['Country Code']=='BGD']
```

```python
#reshape data
inflation_data_melt=inflation_data.melt(
    id_vars=['Country Code','Country Name'],
    value_vars=[str(_) for _ in range(2000,2020) if str(_) in inflation_data.columns],
    var_name='year',
    value_name='rate',
)
inflation_data_melt.rate=inflation_data_melt.rate/100+1
inflation_data_melt
```

```python
#Compute compound rate
compound_inflation_data=pd.DataFrame.from_records([
    {
        'Country Code': code,
        'Compound Rate': group.rate.product()
    } for code, group in inflation_data_melt.groupby('Country Code')
])
pd.options.display.float_format='{:.2f}'.format
compound_inflation_data
```
