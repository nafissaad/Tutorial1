```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from bdshare import get_basic_hist_data
import seaborn as sns
```

```python
stock_names=['ISLAMIBANK', 'BATBC', 'RENATA', 'LINDEBD', 'BERGERPBL', 'BXPHARMA',
       'SQURPHARMA', 'GP', 'OLYMPIC', 'UPGDCL', 'DELTALIFE', 'IBNSINA',
       'WALTONHIL','UNIQUEHRL', 'ADNTEL', 'ITC', 'APEXFOODS', 'FORTUNE',
       'AMANFEED','ACI', 'IFILISLMF1', 'AIBL1STIMF', 'RECKITTBEN',
       'JMISMDL']
end_date=datetime.today()
n_years=5
start_date=end_date-timedelta(days=n_years*365)
df=pd.DataFrame()
for i in stock_names:
    df[i]=get_basic_hist_data(str(start_date),'2023-7-21',i,index='date')['close']
```

```python
df=pd.DataFrame(df['UNIQUEHRL'])
df
```

```python
df['daily_return']=df['UNIQUEHRL'].pct_change()
df['state']=np.where(df['daily_return']>=0,'up','down')
df.head(20)
```

```python
up_counts=len(df[df['state']=='up'])
up_counts=len(df[df['state']=='down'])
up_to_up=len(df[(df['state']=='up')&(df['state'].shift(-1)=='up')])/len(df.query('state=="up"'))
down_to_up=len(df[(df['state']=='up')&(df['state'].shift(-1)=='down')])/len(df.query('state=="up"'))
up_to_down=len(df[(df['state']=='down')&(df['state'].shift(-1)=='up')])/len(df.query('state=="down"'))
down_to_down=len(df[(df['state']=='down')&(df['state'].shift(-1)=='down')])/len(df.query('state=="down"'))
```

```python
transition_matrix=pd.DataFrame({
    'up':[up_to_up,up_to_down],
    'down':[down_to_up,down_to_down]
},index=['up','down'])
```
