```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import seaborn as sn
sn.set()
import datetime as dt
```

```python
start=dt.date.today()-dt.timedelta(500)
start
```

```python
data=pd.DataFrame(pdr.get_data_yahoo('AAPL',start)['Close'])
time=np.arange(1,len(data)+1)
data['time']=time
data=data[['time','Close']]
data=round(data,2)
data.to_csv('C:\\Users\\nafissaad\\Tutorial\\Python for finance\\AAPL.csv')
```

```python
reg=np.polyfit(data['time'][:30],data['Close'][:30],deg=1)
std1=data['Close'][:30].std()
```

```python
trend1=np.polyval(reg,data['time'][:30])
```

```python
def trigger_value(data,Column,index1,index2):
    x=data[Column][index1:index2].min()*1.05
    return x
trigger_value(data,'Close',0,8)
def take_off(data,Column,index1,index2):
    x=data[Column][index1:index2].max()*1.05
    return x
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.array(data['time'][:40]),np.array(data['Close'][:40]),label='S&P500')
plt.plot(np.array(data['time'][:30]),trend1,'r--')
plt.plot(np.array(data['time'][:30]),trend1+std1,'g--')
plt.plot(np.array(data['time'][:30]),trend1-std1,'g--')
plt.axhline(trigger_value(data,'Close',0,30), linestyle='--', color='k')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```
