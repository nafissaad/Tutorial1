```python
import yfinance as yf
import pandas as pd
import datetime as dt
import sqlalchemy
from pandas.tseries.offsets import MonthEnd
```

```python
wiki='https://en.wikipedia.org/wiki/List_of_BSE_SENSEX_companies'
tickersSensex=pd.read_html(wiki)[0].Symbol.to_list()
tickersSensex
```

```python
data=pd.DataFrame()

for ticker in tickersSensex:
    data[ticker]=(yf.download(ticker)['Adj Close'])
```

```python
tickersSensex.index('HDFC.BO')
```

```python
df=data.groupby('Date').sum()
```

```python
df
```

```python
mtlprice=df.resample('M').last()#taking the monthly last Price
```

```python
mtlprice
```

```python
formation=dt.datetime(2011,2,28)
```

```python
begin_measurement=formation-MonthEnd(12)
```

```python
begin_measurement
```

```python
end_measurement=formation-MonthEnd(1)
end_measurement
```

```python
price_end=mtlprice.loc[end_measurement]
price_end
```

```python
price_12monthbefore=mtlprice.loc[begin_measurement]
price_12monthbefore
```

```python
ret_12=price_end/price_12monthbefore-1
```

```python
ret_12#return over the last 12 month, skipping the most recent month
```

```python
winners=ret_12.nlargest(3)
winners
```

```python
winners.index
```

```python
winnerret=mtlprice.loc[formation+MonthEnd(1),winners.index]/mtlprice.loc[formation,winners.index]-1
winnerret
```

```python
Momentumprofit=winnerret.mean()
Momentumprofit
```

```python
def momentumprofit(formation,holdingperiod=1):
    begin_measurement=formation-MonthEnd(12)
    end_measurement=formation-MonthEnd(1)
    price_end=mtlprice.loc[end_measurement]
    price_12mbefore=mtlprice.loc[begin_measurement]
    ret_12=price_end/price_12monthbefore-1
    winners=ret_12.nlargest(3)
    winnerret=mtlprice.loc[formation+MonthEnd(holdingperiod),winners.index]/mtlprice.loc[formation,winners.index]-1
    Momentumprofit=winnerret.mean()
    return Momentumprofit
```

```python
formation
```

```python
momentumprofit(formation)
```

```python
profit,dates,holding=[],[],[]
```

```python
for i in (1,3,6):
    for j in mtlprice.index[12:-i]:
        profit.append(momentumprofit(j,holdingperiod=i))
        dates.append(j + MonthEnd(i))
        holding.append(i)
```

```python
frame=pd.DataFrame({'Momentumprofit':profit,'holdingperiod':holding},index=dates)
frame
```

```python
frame.groupby('holdingperiod').mean()
```
