```python
tickers=df['Symbol'].to_list()
```

```python
yf.download('HDFC.BO')
```

```python
codes=[i[:-3] for i in tickers]
codes
```

```python
india=[]
for i in codes:
    india.append(i+".BO")
```

```python
india
```

```python
frame=yf.download(india)
```

```python
returns=frame['Adj Close'].pct_change()
```

```python
returns=returns.dropna()
```

```python
(1+returns).cumprod().plot(figsize=(10,10))
```
