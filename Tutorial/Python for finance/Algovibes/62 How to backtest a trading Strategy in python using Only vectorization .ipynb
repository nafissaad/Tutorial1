```python
import yfinance as yf 
import pandas as pd
import ta
import numpy as np
import matplotlib.pyplot as plt
```

```python
def indicators(df):
    df['SMA_200']=ta.trend.sma_indicator(df.Close,window=200)
    df['stoch']=ta.momentum.stoch(df.High,df.Low,df.Close,window=10)
    df.dropna(inplace=True)
```

```python
df=yf.download('MMM',start='2010-01-01')
df
```

```python
indicators(df)
```

```python
df
```

```python
df['Buy']=(df.Close>df.SMA_200)&(df.stoch<5)
df
```

```python
df[df.Buy]
```

```python
df['buyprice']=np.where(df.Buy,df.Close*0.97,np.nan)
df.head(50)
```

```python
df.buyprice=df.buyprice.ffill()
df
```

```python
df['sellprice']=df.Open.shift(-1)
df
```

```python
for i in range(1,11):
    df['shifted_low_'+str(i)]=df.Low.shift(-i)
    df['shifted_Close_'+str(i)]=df.Close.shift(-i)
colname_low=['shifted_low_'+str(i) for i in range(1,11)]
colname_close=['shifted_Close_'+str(i) for i in range(1,11)]
```

```python
raw_signals[colname_low].le(raw_signals.buyprice,axis=0)#le means less than or equal to
```

```python
checkbuys=raw_signals[colname_low].le(raw_signals.buyprice,axis=0)
checkbuys
```

```python
checkbuys=checkbuys.astype(int)
checkbuys
```

```python
check_sum=checkbuys.cumsum(axis=1)==1
```

```python
filter_buys=checkbuys[check_sum]
```
