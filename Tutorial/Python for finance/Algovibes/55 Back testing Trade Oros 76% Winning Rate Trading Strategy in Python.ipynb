```python
import yfinance as yf
import pandas as pd
import numpy as np
import ta
```

```python
df=yf.download('BTC-USD',start='2022-01-01',interval='1h')
```

```python
df
```

```python
df['stochrsi_k']=ta.momentum.stochrsi_k(df.Close)
df
```

```python
df['stochrsi_d']=ta.momentum.stochrsi_d(df.Close)
df
```

```python
for i in (8,14,50):
    df['EMA_'+str(i)]=ta.trend.ema_indicator(df.Close,window=i)
```

```python
df
```

```python
df['atr']=ta.volatility.average_true_range(df.High,df.Low,df.Close)
df
```

```python
df.dropna(inplace=True)
df
```

```python
def checkcross(df):
    series=np.where(df['stochrsi_k']>df['stochrsi_d'],1,0)
    df['cross']=series
    df['cross']=df['cross'].diff()
    df['cross']=np.where(df['cross']>0,True,False)
    return df
```

```python
df
```

```python
df['stochrsi_k']>df['stochrsi_d']
```

```python
df=checkcross(df)
```

```python
df# at 2022-01-04 21:00:00+00:00 stochrsi_k > stochrsi_d but that is not the case at 2022-01-04 20:00:00+00:00
```

```python
df['TP']=df.Close+(df.atr*2)
```

```python
df['SL']=df.Close-(df.atr*3)
df
```

```python
df['Buysignals']=np.where((df.cross)&
                          (df.Close>df.EMA_8)
                          &(df.EMA_8>df.EMA_14)
                          &(df.EMA_14>df.EMA_50),1,0)
```

```python
df[df.Buysignals==1]
```

```python
df.iloc[0]
```

```python
df.iloc[0].name
```

```python
selldates=[]
outcome=[]
for i in range(len(df)):
    if df.Buysignals.iloc[i]:
        k=1
        SL=df.SL.iloc[i]
        TP=df.TP.iloc[i]
        in_position=True
        while in_position:
            loop_close=df.Close.iloc[i+k]
            if loop_close>=TP:
                selldates.append(df.iloc[i+k].name)
                outcome.append('TP')
                in_position=False#Escaping the while loop
            elif loop_close<=SL:
                selldates.append(df.iloc[i+k].name)
                outcome.append('SL')
                in_position=False#Escaping the while loop
            k+=1
```

```python
selldates
```

```python
outcome
```

```python
df.loc[selldates,'sellsignals']=1
df.loc[selldates,'outcome']=outcome
df
```

```python
df.sellsignals=df.sellsignals.fillna(0).astype(int)
```
