```python
import yfinance as yf
import pandas as pd
from pandas.tseries.offsets import MonthEnd
```

```python
tickers=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
tickers=tickers['Symbol']
```

```python
df_=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[1]
df_
```

```python
df_.Date=pd.to_datetime(df_.Date.Date)
```

```python
df_
```

```python
df_=df_[df_.Date.Date>='2020-01-01']
df_
```

```python
df_.Added.Ticker
```

```python
tickers=tickers[~(tickers.isin(df_.Added.Ticker))]
tickers
```

```python
ticker_rem=df_.Removed.Ticker
ticker_rem
```

```python
ticker_rem
```

```python
tickers=tickers.append(ticker_rem)
tickers
```

```python
tickers.drop_duplicates(inplace=True)
tickers.dropna(inplace=True)
```

```python
tickers
```

```python
price,symbols=[],[]
start='2020-01-01'
for symbol in tickers:
    df=yf.download(symbol,start=start)['Adj Close']
    if not df.empty:
        price.append(df)
        symbols.append(symbol)
```

```python
price
```

```python
all_prices=pd.concat(price,axis=1)
```

```python
all_prices
```

```python
all_prices.columns=symbols
all_prices
```

```python
all_prices.pct_change()
```

```python
(all_prices.pct_change()+1).prod()-1#Cumulative return
```

```python
all_mtl_ret=all_prices.pct_change().resample('M').agg(lambda x:(x+1).prod()-1)#Monthly cumulative return 
all_mtl_ret
```

```python
all_mtl_ret_12=all_mtl_ret.rolling(12).agg(lambda x: (x+1).prod()-1)
all_mtl_ret_12
```

```python
all_mtl_ret_12.dropna(inplace=True)
all_mtl_ret_12
```

```python
curr_=all_mtl_ret_12.iloc[3]
curr_
```

```python
win_=curr_.nlargest(10)
win_
```

```python
win_ret=all_mtl_ret.loc[win_.name+MonthEnd(1),win_.index]
win_ret
```

```python
win_ret.mean()
```

```python
def mom(all_mtl_ret,all_mtl_ret_12):
    
    ret=[]
    
    for i in range(len(all_mtl_ret_12)-1):
        curr_=all_mtl_ret_12.iloc[i]
        win_=curr_.nlargest(10)
        win_ret=all_mtl_ret.loc[win_.name+MonthEnd(1),win_.index]
        ret.append(win_ret.mean())
    return(ret)

    
```

```python
mom(all_mtl_ret,all_mtl_ret_12)
```

```python
import numpy as np
np.mean(mom(all_mtl_ret,all_mtl_ret_12))
```
