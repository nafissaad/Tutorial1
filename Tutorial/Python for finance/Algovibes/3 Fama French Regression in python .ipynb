```python
factors=reader.DataReader('F-F_Research_Data_Factors','famafrench',start,end)
factors
```

```python
factors=factors[0]#Mkt-RF market risk factor RF risk free
factors=factors[1:]
factors
```

```python
founsret_mtl.shape,factors.shape
```

```python
founsret_mtl.index=factors.index
founsret_mtl
```

```python
merge=pd.merge(founsret_mtl,factors,on='Date')
merge
```

```python
merge[factors.columns]=merge[factors.columns]/100#Scaling percentage to per unit of investment
```

```python
merge['FDGRX']=merge.FDGRX-merge.RF
merge
```

```python
y=merge['FDGRX']
x=merge[merge.columns[1:4]]
X_sm=sm.add_constant(x)
```

```python
model=sm.OLS(y,X_sm)
results=model.fit()
results.summary()
```

```python
#this asset has more risk than market risk because coef MKT-fg more than 1
#on Averahe return is 0.0020 2% monthly return but it is not significant P>0.05 at 95% confident 
#SMB has 0.20 
#overall the model has succeed to explain 93% variation r_squared is 0.93
```
