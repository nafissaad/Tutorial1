```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
assets=['AAPL','MSFT','TSLA','MMM']
```

```python
df=yf.download(assets,start='2020-01-01')
```

```python
df
```

```python
df['Adj Close']
```

```python
prices=df['Adj Close']
prices
```

```python
prices.plot()
```

```python
returns=prices.pct_change()
returns
```

```python
cum_ret=(returns+1).cumprod()
cum_ret
```

```python
cum_ret.plot(figsize=(10,6))
```

```python
returns.std()
```

```python
counter=0
fig,ax=plt.subplots(2,2,figsize=(19,10))
for i in range(2):
    for j in range(2):
        ax[i,j].plot(returns[returns.columns[counter]])
        ax[i,j].set_title(returns.columns[counter])
        ax[i,j].set_xticklabels([i.year for i in returns.index], rotation=45)
        counter+=1
        #print(i,j)
```
