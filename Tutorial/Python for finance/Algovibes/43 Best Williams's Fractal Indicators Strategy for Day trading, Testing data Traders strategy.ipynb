```python
import yfinance as yf
import numpy as np
import pandas as pd
import ta
```

```python
df=yf.download('TSLA',start='2022-11-21',interval='15m')
df
```

```python
df['EMA200']=ta.trend.ema_indicator(df.Close,window=200)
```

```python
df
```

```python
df[:5]
```

```python
df[:5].High.rolling(3,center=True).max()#2022-11-30 13:30:00-05:00	182.489899	185.300003	181.550003	184.625504	184.625504	5550655	NaN
```

```python
df['wf_Top_bool']=np.where(df['High']==df['High'].rolling(9,center=True).max(),True,False)
```

```python
df.head(30)
```

```python
df['wf_Top']=np.where(df['High']==df['High'].rolling(9,center=True).max(),df['High'],None)
df
```

```python
df['wf_Top']=df['wf_Top'].ffill()
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['Buy']=np.where((df.Close > df.wf_Top) & (df.Close>df.EMA200),1,0)
```

```python
df
```

```python
df['SL']=np.where(df.Buy==1,df.Close-(df.Close-df.Low),0)
df['TP']=np.where(df.Buy==1,df.Close+(df.Close-df.Low)*1.5,0)
```

```python
buys,sell=[],[]
```

```python
for i in range(len(df)):
    if df.Buy.iloc[i]:#if buy condition is True
        buys.append(df.iloc[i].name)
        for j in range(len(df)-i):
            if df.TP.iloc[i]<df.Close.iloc[i+j] or \
            df.SL.iloc[i]>df.Close.iloc[i+j]:
                sell.append(df.iloc[i+j].name)
                break
```

```python
buys
```

```python
sell
```

```python
frame=pd.DataFrame([buys,sell])
frame
```

```python
frame=frame.T
frame
```

```python
frame=frame.dropna()
frame
```

```python
frame.columns=['Buys','Sells']
frame
```

```python
actualtrades=frame[frame.Buys>frame.Sells.shift(1)]
actualtrades
```

```python
profits=(df.loc[actualtrades.Sells].Close.values-\
        df.loc[actualtrades.Buys].Close.values)/df.loc[actualtrades.Buys].Close.values
```

```python
profits
```
