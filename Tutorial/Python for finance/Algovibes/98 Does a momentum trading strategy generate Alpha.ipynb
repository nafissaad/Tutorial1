```python
df=pd.read_excel('C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\My Portfolio\\portfolio_performence.xlsx')
df.rename(columns={'Unnamed: 0':'date'},inplace=True)
df.drop(0,inplace=True)
df.set_index('date',inplace=True)
```

```python
s_p_mtl=(df['Total'].pct_change()+1).resample('M').apply(np.prod)-1
len(s_p_mtl)
```

```python
s_p_mtl.name='Portfolio'
```

```python
s_p_mtl
```

```python
tog=pd.concat([s_p_mtl,mom_ret],axis=1)
tog
```

```python
tog=tog.dropna()
len(tog)
```

```python
(tog+1).cumprod().plot()
```

```python
tog['rf']=[i for i in np.linspace(0.068, 0.075, len(tog))]#for time constraint creating rf data
```

```python
tog
```

```python
tog['Month']=[i+1 for i  in range(len(tog))]
tog
```

```python
tog['r-rf']=tog['MOM']-tog['rf']
tog['rm-rf']=tog['Portfolio']-tog['rf']
```

```python
tog=tog.assign(Index=range(len(tog))).set_index('Index')
tog
```

```python
import statsmodels.api as sm
```

```python
x=tog['rm-rf']
```

```python
x=sm.add_constant(x)
```

```python
x
```

```python
y=tog['r-rf']
```

```python
model=sm.OLS(y,x)
```

```python
results=model.fit()
```

```python
results.summary()
```
