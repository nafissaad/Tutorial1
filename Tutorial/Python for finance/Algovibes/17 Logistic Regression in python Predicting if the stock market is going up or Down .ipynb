```python
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
```

```python
data=yf.download("^GSPC",start='2001-01-03',end='2005-12-31')
```

```python
data
```

```python
df=data['Adj Close'].pct_change()*100
```

```python
df=df.rename('Today')
```

```python
df=df.reset_index()
df
```

```python
for i in range(1,6):
    df['Lag'+str(i)]=df['Today'].shift(i)
```

```python
df
```

```python
df['volume']=data.Volume.shift(1).values/1000000000
```

```python
df
```

```python
df=df.dropna()
```

```python
df
```

```python
df['Direction']=[1 if i>0 else 0 for i in df['Today']]
```

```python
df
```

```python
df=sm.add_constant(df)
df
```

```python
df.columns.to_list()[0:9]
```

```python
x=df[['const', 'Today', 'Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'volume']]
x
```

```python
y=df.Direction
y
```

```python
model=sm.Logit(y,x)
```

```python
result=model.fit()
```

```python
result.summary()
```

```python
prediction=result.predict(x)
```

```python
prediction
```

```python
y
```

```python
def confusion_matric(act,pred):
    predtrans=['Up' if i > 0.5 else 'Down' for i in pred]
    actuals=['Up' if i > 0.5 else 'Down' for i in act]
    confusion_matrix=pd.crosstab(pd.Series(actuals),
                                pd.Series(predtrans),
                                rownames=['Actual'],
                                colnames=['Predicted'])
    return confusion_matrix
```

```python
confusion_matric(y,prediction)
```

```python
x_train=df[df.Date.dt.year<2005][['const', 'Today', 'Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'volume']]
y_train=df[df.Date.dt.year<2005]['Direction']
x_test=df[df.Date.dt.year==2005][['const', 'Today', 'Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'volume']]
y_test=df[df.Date.dt.year==2005]['Direction']
```

```python
model=sm.Logit(y_train,x_train)
```

```python
result=model.fit()
```

```python
prediction=result.predict(x_test)
```

```python
confusion_matric(y_test,prediction)
```
