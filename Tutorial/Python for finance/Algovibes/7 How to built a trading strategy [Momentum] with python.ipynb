```python
import pandas as pd
import pandas_datareader.data as reader
import datetime as dt
from bdshare import get_basic_hist_data
from dateutil.relativedelta import relativedelta
```

```python
#Get ticker symbol for the stocks cntained in DJI
table=pd.read_html('https://www.amarstock.com/latest-share-price')[0]
table
```

```python
table.to_csv('E:\\BD_stocks.csv')
```

```python
tickers=table['Trading code'].tolist()
tickers
```

```python
#get price for the DJI components
start=dt.datetime(2020,1,1)
end=dt.datetime(2022,8,10)
```

```python
def get_close_price(tickers):
    df=pd.DataFrame()
    x=1
    for i in tickers[0:225]:
        print('working on :',i,'Tickert number :',x) 
        df[i]=get_basic_hist_data(start,end,i,index='date')['close']
        print('Colleted :',i)
        x=x+1
    return df
```

```python
df=get_close_price(tickers)
type(df.index)
```

```python
df.index=pd.to_datetime(df.index)
type(df.index)
```

```python
#calculationg monthly returns by calculating daily return
mtl_ret=df.pct_change().resample('M').agg(lambda x:(x+1).prod()-1)
mtl_ret
```

```python
#calculate returns over the past 11 month
import numpy as np
past_11=(mtl_ret+1).rolling(11).apply(np.prod)-1
past_11
```

```python
mtl_ret+1
```

```python
from datetime import timedelta

formation=dt.datetime(2022,6,30)
end_measurement=formation - relativedelta(months=1)
end_measurement=end_measurement+timedelta(1)
end_measurement
```

```python
past_12=past_11.loc[end_measurement]
past_12=past_12.reset_index()
past_12
```
