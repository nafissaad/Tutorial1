```python
import yfinance as yf
import pandas as pd
import ta
```

```python
df=yf.download('NFLX',start='2010-01-01')
df
```

```python
df['SMA']=df.Close.rolling(200).mean()
df
```

```python
df['rsi']=ta.momentum.rsi(df.Close,window=10)
df
```

```python
df['price']=df.Open.shift(-1)
df
```

```python
df['Buy']=(df.Close>df.SMA)&(df.rsi<30)
df[df.Buy]
```

```python
df['Sell']=df.rsi>40
df
```

```python
in_position=False
buys,sells=[],[]
buydates,selldates=[],[]
for index,row in df.iterrows():
    if not in_position:
        if row.Buy:
            buys.append(row.price)
            in_position=True
            r_count=0
    if in_position:
        r_count+=1
        if row.Sell:
            sells.append(row.price)
            in_position=False
        elif r_count>=10:
            sells.append(row.price)
            in_position=False
```

```python
sells
```

```python
trades=pd.DataFrame([buys,sells],index=['Buys','Sells'])
trades
```

```python
trades=trades.T
trades
```

```python
trades['PnL']=(trades.Sells-trades.Buys)/trades.Buys
trades
```

```python
trades[trades.PnL>0].shape[0]/trades.shape[0]
```
