```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
Ticker_df=pd.read_html('https://en.wikipedia.org/wiki/Nasdaq-100')[4]
Ticker_df
```

```python
tickers=Ticker_df.Ticker.to_list()
tickers
```

```python
df=yf.download(tickers,start='2010-01-01')['Adj Close']
df
```

```python
df=df.dropna(axis=1)
df
```

```python
df.pct_change()
```

```python
mtl=(df.pct_change()+1)[1:].resample('M').prod()#Monthly_return
```

```python
mtl
```

```python
def get_rolling_ret(df,n):
    return df.rolling(n).apply(np.prod)
```

```python
get_rolling_ret(mtl,12)
```

```python
ret_12,ret_6,ret_3=get_rolling_ret(mtl,12),get_rolling_ret(mtl,6),get_rolling_ret(mtl,3)
```

```python
ret_6
```

```python
ret_12.loc['2010-12-31']
```

```python
ret_12.loc['2010-12-31'].nlargest(50)
```

```python
top_50=ret_12.loc['2010-12-31'].nlargest(50).index
```

```python
ret_6.loc['2010-12-31',top_50]
```

```python
top_30=ret_6.loc['2010-12-31',top_50].nlargest(30).index
```

```python
top_30
```

```python
top_10=ret_3.loc['2010-12-31',top_30].nlargest(10).index
top_10
```

```python
def get_top(date):
    top_50=ret_12.loc['2010-12-31'].nlargest(50).index
    top_30=ret_6.loc['2010-12-31',top_50].nlargest(30).index
    top_10=ret_3.loc['2010-12-31',top_30].nlargest(10).index
    return top_10
```

```python
get_top('2010-12-31')
```

```python
mtl.loc['2010-12-31':][1:2]
```

```python
portfolio=mtl.loc['2010-12-31':,get_top('2010-12-31')][1:2]
```

```python
portfolio
```

```python
portfolio.mean(axis=1)
```

```python
portfolio.mean(axis=1).values[0]
```

```python
def pf_performance(date):
    portfolio=mtl.loc[date:,get_top(date)][1:2]
    return portfolio.mean(axis=1).values[0]
```

```python
pf_performance('2010-12-31')
```

```python
mtl
```

```python
mtl.index
```

```python
returns=[]
for date in mtl.index[:-1]:
    returns.append(pf_performance(date))
```

```python
returns
```

```python
pd.Series([i-0.01 for i in returns],index=mtl.index[1:]).cumprod().plot()
```
