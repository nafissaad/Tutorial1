```python
import yfinance as yf
```

```python
df=yf.download('^GSPC',start='1996-01-01',end='2020-11-01')
#df.reset_index(drop=True, inplace=True)#to drom index columns
df
```

```python
df['MA200']=df['Adj Close'].rolling(window=200).mean()
df
```

```python
df=df.dropna()
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Adj Close'])
plt.plot(df['MA200'])
plt.show()
```

```python
import pandas as pd
pd.options.mode.chained_assignment=None#turn off error massage
```

```python
df['price change']=df['Adj Close'].pct_change()
```

```python
df=df.dropna()
df
```

```python
df['Upmove']=df['price change'].apply(lambda x: x if x > 0 else 0)
df['Downmove']=df['price change'].apply(lambda x: abs(x) if x < 0 else 0)
df
```

```python
df['price change'][0:10].apply(lambda x: x if x < 0 else 0),df['price change'][0:10].apply(lambda x: abs(x) if x < 0 else 0)
```

```python
df['avg Up']=df['Upmove'].ewm(span=19).mean()
df['avg Down']=df['Downmove'].ewm(span=19).mean()
df
```

```python
df['RS']=df['avg Up']/df['avg Down']
df
```

```python
df['RSI']=df['RS'].apply(lambda x: 100-(100/(x+1)))
df
```

```python
df.loc[(df['Adj Close']>df['MA200']) & (df['RSI']<30),'Buy']='Yes'
```

```python
df.loc[(df['Adj Close']<df['MA200']) | (df['RSI']>30),'Buy']='No'
```

```python
df
```

```python
profit_or_loss=[]
for i in range(len(df)-13):#limiting the loop from going empty space at the end
    if 'Yes' in df['Buy'].iloc[i]:
        for j in range(1,12):#your selling condition is RSI is not > 40 after 10 trading day
            if df['RSI'].iloc[i+j]>40:
                profit_or_loss.append(df['Open'].iloc[i+j+1]-df['Open'].iloc[i+1])#+1 for next day price after the event, buy after the buy signal  
                break
            if df['RSI'].iloc[i+j]<40:
                profit_or_loss.append(df['Open'].iloc[i+13]-df['Open'].iloc[i+1])#After 12 trading days i am gonna sell
                break      
```
