```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df=yf.download('AAPL',start='2023-02-02',interval='1H')
df.head(50)
```

```python
def get_levels(date):
    values=[-0.618,0.618,1.618]
    series_=df.loc[date:][1:2].squeeze()
    diff=series_.High-series_.Low
    levels=[series_.Close+i*diff for i in values]
    return levels
```

```python
get_levels('2022-01-01')
```

```python
df.index.date
```

```python
dates=np.unique(df.index.date)
dates
```

```python
Date='2022-01-01'
df[Date:][2:]
```

```python
buys,sells=[],[]
trade_dates=[]
in_position=False
for date in dates:
    for index,row in df[date:][2:].iterrows():
        if not in_position:
            sl,entry,tp=get_levels(date)
            if row.Close>=entry:
                buys.append(row.Close)
                trade_dates.append(date)
                in_position=True
        elif in_position:
            if row.Close>=tp or row.Close <=sl:
                sells.append(row.Close)
                in_position=False
            break
```

```python
trades=pd.DataFrame([buys,sells])
```

```python
trades
```

```python
trades.columns=trade_dates
trades=trades
trades.index=['Buy','Sell']
trades
```

```python
trades=trades.T
trades
```

```python
trades['PnL']=trades.Sell-trades.Buy
trades['Pnl_rel']=trades.PnL/trades.Buy
```

```python
(trades.Pnl_rel+1).cumprod()
```
