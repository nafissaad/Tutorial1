```python
def getminutedata(symbol,interval,lookback):
    frame=pd.DataFrame(yf.download(symbol,period=interval,interval=lookback))
    #frame=frame.iloc[:,:6]
    #frame.columns=['T','O','H','L','C','A']
    return frame
    
```

```python
def strategytest(symbol,qty,entried=False):
    df=getminutedata(symbol,'1','1m')
    cumulret=(df.Open.pct_change()+1).cumprod()-1
    c=cumulret.index[-1]
    if not entried:#when entried condition is False
        if cumulret[-1]<-0.002:
            print(c)
            print("Buy")
            entried=True
        else:
            print('No trade has been executed')
    if entried:#when entried condition is false
        while True:
            df=getminutedata(symbol,'1','1m')
            sincebuy=df[c:df.index[-1]]
            cumulret=(sincebuy.Open.pct_change()+1).cumprod()-1
            if cumulret[-1]>0:
                print(df.index[-1])
                print(cumulret[-1])
                print("Sell")
                entried=False
                break
```

```python
strategytest('BTC-USD',4)
```
