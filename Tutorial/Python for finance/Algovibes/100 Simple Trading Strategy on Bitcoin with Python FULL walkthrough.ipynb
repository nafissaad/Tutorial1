```python
import pandas as pd
from binance.client import Client
import numpy as np
client=Client()
```

```python
def getdata(symbol,start):
    frame=pd.DataFrame(client.get_historical_klines(symbol,'1h',start))
    frame=frame.iloc[:,:6]
    frame.columns=['Time','Open','High','Low','Close','Volume']
    frame.set_index('Time',inplace=True)
    frame.index=pd.to_datetime(frame.index,unit='ms')
    frame=frame.astype(float)
    return frame
```

```python
df=getdata('BTCUSDT','2020-01-01')
df
```

```python
df['chg']=np.log(df.Close.pct_change()+1)
df
```

```python
df['12h_chg']=df.chg.rolling(12).sum()
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['buyprice']=df.Open.shift(-1)
df
```

```python
df
```

```python
def profcalc(change,tp,sl):
    in_position=False

    profits=[]

    for index,row in df.iterrows():
        if not in_position:
            if row['12h_chg']>change:
                buyprice=row.buyprice
                in_position=True
        if in_position:
            if row.High>= buyprice*tp:
                sellprice=buyprice*tp
                profit=(sellprice-buyprice)/buyprice
                profits.append(float(profit))
                in_position=False
                
            elif row.Low<=buyprice*sl:
                sellprice=buyprice*sl
                profit=(sellprice-buyprice)/buyprice
                profits.append(float(profit))
                in_position=False
    return((pd.Series(profits)+1).prod()-1)
```

```python
profcalc(0.02,1.07,0.93)
```

```python
arr=np.arange(0.01,0.16,0.01)
arr
```

```python
for ele in arr:
    print('for '+str(ele))
    print(profcalc(0.02,1+ele,1-(ele/2)))
```
