```python
import yfinance as yf
import numpy as np
import pandas as pd
import datetime as dt
from pandas.tseries.offsets import MonthEnd
```

```python
df=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
df
```

```python
tickers=df.Symbol.to_list()
tickers
```

```python
df2=yf.download(tickers,starts='2009-12-31',end='2020-12-31')
df2
```

```python
price=df2['Adj Close']
price
```

```python
price.index=pd.to_datetime(price.index)
```

```python
Monthly_ret=price.pct_change().resample('M').agg(lambda x: (x+1).prod()-1)
```

```python
past_11=(Monthly_ret+1).rolling(11).apply(np.prod)-1
past_11
```

```python
formation=dt.datetime(2010,12,31)
```

```python
formation
```

```python
end_measure=formation-MonthEnd(1)
end_measure
```

```python
past_12=past_11.loc[end_measure]
past_12
```

```python
past_12=past_12.reset_index()
```

```python
past_12
```

```python
past_12['Decile']=pd.qcut(past_12.iloc[:,1],10,labels=False,duplicates='drop')
```

```python
past_12
```

```python
winners=past_12[past_12.Decile==9]
loser=past_12[past_12.Decile==0]
```

```python
winners
```

```python
winnerret=Monthly_ret.loc[formation+MonthEnd(1),Monthly_ret.columns.isin(winners['index'])]
loserret=Monthly_ret.loc[formation+MonthEnd(1),Monthly_ret.columns.isin(loser['index'])]
```

```python
momentumprofit=winnerret.mean()-loserret.mean()
momentumprofit
```

```python
def momentum(formation):
    end_measure=formation-MonthEnd(1)
    past_12=past_11.loc[end_measure]
    past_12=past_12.reset_index()
    past_12['Decile']=pd.qcut(past_12.iloc[:,1],10,labels=False,duplicates='drop')
    winnerret=Monthly_ret.loc[formation+MonthEnd(1),Monthly_ret.columns.isin(winners['index'])]
    loserret=Monthly_ret.loc[formation+MonthEnd(1),Monthly_ret.columns.isin(loser['index'])]
    momentumprofit=winnerret.mean()-loserret.mean()
    return momentumprofit
```

```python
momentum(formation)
```

```python
for i in range(12*10):
    print(formation+MonthEnd(i))
```

```python
profits=[]
dates=[]
for i in range(12*10):
    profits.append(momentum(formation+MonthEnd(i)))
    dates.append(momentum(formation+MonthEnd(i)))
```
