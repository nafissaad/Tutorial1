```python
import pandas_datareader.data as reader
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
```

```python
end=dt.datetime.now()
start=dt.date(end.year-1,end.month,end.day)
```

```python
portfolio=['GOOG','AAPL','MSFT','TSLA','^GSPC','^BSESN']#BSESN is indian scok exchange index
```

```python
df=reader.get_data_yahoo(portfolio,start,end)['Adj Close']
df
```

```python
returns=df.pct_change()
```

```python
returns.cov()#possitive covariance means possitive dircectional change 
```

```python
returns.var()#covariance with itself is just variance 
```

```python
returns.corr()
```

```python
sns.heatmap(returns.corr());
```
