```python
dji=pd.read_html('https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')[1]
dji
```

```python
tickers=dji.Symbol.to_list()
```

```python
frames=[]
for ticker in tickers:
    print(ticker)
    var=yf.Ticker(ticker)
    frame=var.institutional_holders
    frame['comp']=var.ticker
    frames.append(frame)
```

```python
frames
```

```python
all_tog=pd.concat(frames)
all_tog
```

```python
G=nx.from_pandas_edgelist(all_tog,'Holder','comp',edge_attr=True)
```

```python
edgelist=nx.to_edgelist(G)
```

```python
G.nodes
```

```python
colors=[]
for node in G:
    if node in all_tog['comp'].values:
        colors.append('red')
    else:
        colors.append('yellow')
```

```python
#making the node size depended on degree
plt.figure(figsize=(20,20))
nx.draw(G,with_labels=True,
       node_color=colors,
       node_size=[v*100 for v in dict(G.degree()).values()],
       width=[v[2]['Value']/50_000_000_000 for v in edgelist])
```
