```python
tickers=['^GSPC']
```

```python
df=yf.download(tickers,start='2014-12-31',end='2022-08-31')
```

```python
df['Return']=df['Close'].pct_change()
df
```

```python
returns=df['Return'].mean()
Volatility=df['Return'].std()
returns,Volatility
```

```python
df1=pd.DataFrame()
df1['Confidence Interval']=np.linspace(0.001,0.05,num=50,dtype='float')
```

```python
df1['VaR(normal)']=[np.nan for i in range(len(df1))]
df1['VaR(average)']=[np.nan for i in range(len(df1))]
df1['VaR(Mill)']=[np.nan for i in range(len(df1))]
df1
```

```python
for i in df1.index:
    df1.loc[i,'VaR(normal)']=returns+norm.ppf(df1.loc[i,'Confidence Interval'])*Volatility#Volatility
    df1.loc[i-1,'VaR(average)']=np.mean(df1['VaR(normal)'].iloc[0:i])
    df1.loc[i,'VaR(Mill)']=returns-Volatility*(norm.pdf((df1.loc[i,'VaR(normal)']-returns)/Volatility)/norm.cdf((df1.loc[i,'VaR(normal)']-returns)/Volatility))
```

```python
#pdf=measures probability density function (pdf) of the distribution
#cdf=Cumulative Density Function (CDF) for a Normal random variable at a value x
#Mills ration=daily returns-standard deviation*(pdf(VaR normal-daily returns)*standard deviation/cdf(VaR normal-daily returns)*standard deviation)
df1
```
