```python
df=pd.read_excel('C:\\Users\\nafissaad\\Downloads\\NEDL_Anderson_Darling.xlsx',index_col='Date')[['S&P 500 return','Unnamed: 8']]
df.to_excel('C:\\Users\\nafissaad\\Downloads\\NEDL_Anderson_Darling.xlsx')
df
```

```python
number_of_observation=len(df)
mean=df['S&P 500 return'].mean()
standard_deviation=df['S&P 500 return'].std()
number_of_observation,mean,standard_deviation
```

```python
df['Sorted']=df['S&P 500 return'].sort_values(ascending=True).values
df
```

```python
df['Rank']=[i for i in range(1,len(df)+1)]
df
```

```python
df['Emperical Distribution']=df['Rank']/len(df['S&P 500 return'])
df
```

```python
n,minmax,mean,var,skew,kurt=stats.describe(df['S&P 500 return'])
mini,maxi=minmax
std=var**0.5#0.0002496130**0.5=0.015799145546516114
overly=np.linspace(mini,maxi,n)
df['Theoritical distribution']=norm.cdf(overly,mean,std)
```

```python
df['Theoretical distribution (backwards)']=df['Theoritical distribution'][::-1].values
df
```

```python
df['S']=(2*df['Rank']-1)/number_of_observation*(np.log(df['Theoritical distribution'])+np.log(1-df['Theoretical distribution (backwards)']))
df
```

```python
S=sum(df['S'])
Anderson_Darling_squared=-number_of_observation-S
Anderson_Darling_Statistics=np.sqrt(Anderson_Darling_squared)
Anderson_Darling_Statistics
```
