```python
param=pd.DataFrame(columns=[['Normal','Logistic','Hyperbolic Scecant','Lance','Cauchy']],index=['Location(Mean)','Scale(std)'])
param.loc['Location(Mean)']=[0.0004,0.0004,0.0004,0.0005,0.0005]
param.loc['Scale(std)']=[0.0085,0.0047,0.0085,0.0058,0.0036]
param
```

```python
pi=3.141592653589793#PI: 3.141592653589793
matrix['Normal']=[param.loc['Location(Mean)','Normal'][0]+param.loc['Scale(std)','Normal'][0]*norm.ppf(i) for i in matrix.index]
matrix['Logistic']=[param.loc['Location(Mean)','Logistic'][0]+param.loc['Scale(std)','Logistic'][0]*np.log(i/(1-i)) for i in matrix.index]
matrix['Hyperbolic Scecant']=[param.loc['Location(Mean)','Hyperbolic Scecant'][0]+2*param.loc['Scale(std)','Hyperbolic Scecant'][0]/pi*np.log(math.tan(pi/2*i)) for i in matrix.index]
matrix['Lance']=[param.loc['Location(Mean)','Lance'][0]+param.loc['Scale(std)','Lance'][0]*np.log(2*i) for i in matrix.index]
matrix['Cauchy']=[param.loc['Location(Mean)','Cauchy'][0]+param.loc['Scale(std)','Cauchy'][0]*math.tan(pi*i-pi/2) for i in matrix.index]
matrix
```
