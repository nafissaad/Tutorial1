```python
import random
import numpy as np
import yfinance as yf
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import norm
import math
import scipy.stats as stats
%matplotlib inline
```

```python
df=yf.download('BLK',start='2017-08-31',end='2022-08-31')
```

```python
df['Retuen']=df['Close'].pct_change()
df
```

```python
from scipy.stats import norm
```

```python
Sample_size=len(df)
Confidence_alpha=0.05
Daily_return=((1+df['Retuen']).prod())**(1/Sample_size)-1
Volatility=df['Retuen'].std()
Var=norm.ppf(Confidence_alpha,Daily_return,Volatility)
Var
```

```python
Daily_return+norm.ppf(Confidence_alpha)*Volatility
```

```python
df['Violation']=[1 if v<Var else 0 for v in df['Retuen'].values]
df
```

```python
proportion_of_violations=df['Violation'].sum()/Sample_size
proportion_of_violations
```

```python
difference=proportion_of_violations-Confidence_alpha
standard_error=math.sqrt((Confidence_alpha*(1-Confidence_alpha))/Sample_size)
z_stat=difference/standard_error
pvalue=2*(1-(stats.norm.cdf(abs(z_stat))))
pvalue#pvalue<5 means our var is significant at given cofident interval,we got enoght evidence
#it is over estimating or under estimating 
```
