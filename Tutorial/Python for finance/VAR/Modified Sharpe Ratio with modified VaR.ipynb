```python
import random
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math
import scipy.stats as stats
from scipy.stats import norm
from scipy.stats import kurtosis
import requests
%matplotlib inline
```

```python
df=pd.read_excel('C:\\Users\\nafissaad\\Downloads\\NEDL_Modified_Sharpe_MVaR.xlsx',index_col='Date')[['Large stocks','Small stocks','Treasury bonds','Corporate bonds','Gold','REIT']]
#df.to_excel('C:\\Users\\nafissaad\\Downloads\\NEDL_PMPT.xlsx')
df
```

```python
returns=df.pct_change().fillna(0)
returns
```

```python
matrix=pd.DataFrame(columns=[['Large Stocks','Small stocks','Treasury bonds','Corporate bonds','Gold','REIT']],index=['Portfolio weights'])
matrix
```

```python
matrix.loc['Portfolio weights']=1/6
matrix
```

```python
df['portfolio']=matrix.loc['Portfolio weights'].dot(returns[returns.columns].values.T)
df
```

```python
returns_mean=(np.prod(1+df['portfolio'].values))**(1/1259)-1#Geometric Mean
volatility_sigma=df['portfolio'].std()
skewness=df['portfolio'].skew()
kurtosises=kurtosis(df['portfolio'])
kurtosises
```

```python
confidence_interval=0.99#less confidence interval means less risk averse
```

```python
z_stat=norm.ppf(0.99)
```

```python
#Favre and Galeano(2002) formula it adjust the normalities and calulate the risk under non normal distribution assumption
Z_modified_value_at_risk=z_stat+1/6*(z_stat**2-1)*skewness+1/24*(z_stat**3-3*z_stat)*kurtosises-1/36*(2*z_stat**3-5*z_stat)*skewness**2
```

```python
Z_modified_value_at_risk
```

```python
MVAR=returns_mean-volatility_sigma*Z_modified_value_at_risk
MVAR
```

```python
annualised_expected_return=(1+returns_mean)**252-1
annualised_expected_return
```

```python
risk_free_rate=0.0157
modified=(annualised_expected_return-risk_free_rate)/(risk_free_rate-MVAR)
modified#it means we gain 1.3% return for 1% modified value at risk
```
