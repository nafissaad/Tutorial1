```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel('C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\Technical analysis\\NEDL_Omega_Ratio.xlsx',index_col='Date')
df=df.rename(columns=df.iloc[0])
df=df[1:]
df=df.iloc[:,0:5]
df
```

```python
returns=df.pct_change()
returns
```

```python
Matrix=pd.DataFrame(columns=[df.columns],index=['Returns','Weights'])
Matrix
```

```python
for i in df.columns:
    Matrix.loc['Returns',i]=np.prod(1+returns[i])**(1/len(df))-1
    Matrix.loc['Weights',:]=0.2
Matrix    
```

```python
Annual_target_return=0.1
Expected_returns=Matrix.loc['Returns'].mean()
Targeted_daily_return=(1+0.1)**(1/252)-1
Outperformance=Expected_returns-Targeted_daily_return
Outperformance
```

```python
returns['Portfolio']=Matrix.loc['Weights'].values.dot(returns.T)
returns
```

```python
returns['Downside']=[max(Targeted_daily_return-i,0) for i in returns['Portfolio'].fillna(1).values]
returns
```

```python
Downside_risk=returns['Downside'].mean()
Downside_risk
```

```python
Omega_rato=(Outperformance/Downside_risk)+1
```
