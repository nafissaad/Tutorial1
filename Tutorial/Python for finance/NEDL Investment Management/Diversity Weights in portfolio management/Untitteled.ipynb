```python
df.columns=df.iloc[0].values
df=df.drop(df.index[0])
df
```

```python
df['Weight']=df['Weight'].astype('float64')/100
df
```

```python
df.info()
```

```python
df1=df[['Company','Weight']]
df1
```

```python
df1['Equal Weights']=[100/len(df1) for i in range(0,len(df1))]
df1
```

```python
p=0.75#When i want 75% diversification
n=len(df1)
Sum=sum(df['Weight']**p)
p,Sum,n
```

```python
Diversified_Weight=[]
for i in df1.index:
    x=(df1.loc[i]['Weight']**p)/Sum
    Diversified_Weight.append(x)
Diversified_Weight
```

```python
df1['Diversified Weight']=Diversified_Weight
df1
```
