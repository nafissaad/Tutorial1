```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel('C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\Technical analysis\\NEDL_Information_Ratio.xlsx',index_col='Date')
df=df.rename(columns=df.iloc[0])
df=df[1:]
df=df.iloc[:,0:4]
df
```

```python
returns=df.pct_change()
returns
```

```python
out_performance=returns.sub(returns['SPY'],axis=0)
out_performance
```

```python
matrix=pd.DataFrame(columns=[returns.columns],index=['Annualised return','Outperformance','Tracking error','Information Ratio'])
matrix.loc['Annualised return','SPY']=np.prod(1+returns['SPY'])**(252/1400)-1
matrix.loc['Annualised return','BLK']=np.prod(1+returns['BLK'])**(252/1400)-1
matrix.loc['Annualised return','SCHW']=np.prod(1+returns['SCHW'])**(252/1400)-1
matrix.loc['Annualised return','QQQ']=np.prod(1+returns['QQQ'])**(252/1400)-1
matrix.loc['Outperformance','BLK']=matrix.loc['Annualised return','BLK'][0]-matrix.loc['Annualised return','SPY'][0]
matrix.loc['Outperformance','SCHW']=matrix.loc['Annualised return','SCHW'][0]-matrix.loc['Annualised return','SPY'][0]
matrix.loc['Outperformance','QQQ']=matrix.loc['Annualised return','QQQ'][0]-matrix.loc['Annualised return','SPY'][0]
matrix.loc['Tracking error','BLK']=out_performance['BLK'].std()*np.sqrt(252)
matrix.loc['Tracking error','SCHW']=out_performance['SCHW'].std()*np.sqrt(252)
matrix.loc['Tracking error','QQQ']=out_performance['QQQ'].std()*np.sqrt(252)
matrix.loc['Information Ratio','BLK']=matrix.loc['Outperformance','BLK'][0]/matrix.loc['Tracking error','BLK'][0]
matrix.loc['Information Ratio','SCHW']=matrix.loc['Outperformance','SCHW'][0]/matrix.loc['Tracking error','SCHW'][0]
matrix.loc['Information Ratio','QQQ']=matrix.loc['Outperformance','QQQ'][0]/matrix.loc['Tracking error','QQQ'][0]
matrix
```

```python
#Information Ratio 0.5 is good....Information Ratio tells us the portfolio is outperforming the market with less deviation 
```
