```python
df=pd.read_excel('C:\\Users\\nafissaad\\Downloads\\NEDL_PMPT.xlsx',skiprows=[0],index_col='Unnamed: 0')[['JPM','WMT','MSFT','KO','PFE','CAT','SP 500']]
#df.to_excel('C:\\Users\\nafissaad\\Downloads\\NEDL_PMPT.xlsx')
df
```

```python
#CShalit and Yitzhaki 1984 showed GINI cofficient can be generalized as trated as measure of risk
returns=df.pct_change().fillna(0)
returns
```

```python
matrix=pd.DataFrame(columns=[['JPM','WMT','MSFT','KO','PFE','CAT']],index=['Returns,historical'])
matrix
```

```python
matrix.loc['Returns,historical']=[((df[i][-1]/df[i][0])**(1/10)-1) for i in df.columns[:-1]]
matrix
```

```python
matrix.loc['Weights']=1/len(df.columns[:-1])
matrix
```

```python
df['Portfolio']=matrix.loc['Weights'].values.dot(returns[df.columns[:-1]].values.T)
df
```

```python
df['Rank']=[i for i in range(len(df))]
df
```

```python
df['Sorted Portfolio']=df['Portfolio'].sort_values(ascending=True).values
df
```

```python
covariance=np.cov([i for i in df['Rank'].values],[i for i in df['Sorted Portfolio'].values])[0][1]
count=len(df)
Historical_average=df['Portfolio'].mean()
Expected_return=matrix.loc['Weights'].values.dot(matrix.loc['Returns,historical'].values.T)
GINI_Index=2*covariance/(Historical_average*count)
covariance,count,Historical_average,Expected_return,GINI_Index
```

```python
#Notice In Economics we have seen that GINI Coefficient Lies betwwen 0 and 1,Because here we have negative return distribution we can have 1<
GINI_Index  =[]
weights=[]
for x in range(80000):
    p_weights=np.random.random(len(df.columns[:-4]))
    p_weights/=np.sum(p_weights)
    Portfolio=p_weights.dot(returns[df.columns[:-4]].T)
    Rank=[i for i in range(len(df))]
    Sorted_Portfolio=np.sort(Portfolio)
    covariance=np.cov(Rank,Sorted_Portfolio)[0][1]
    count=len(df)
    Historical_average=Portfolio.mean()
    Expected_return=p_weights.dot(matrix.loc['Returns,historical'].values.T)
    GINI_Index_1=2*covariance/(Historical_average*count)
    weights.append(p_weights)
    GINI_Index.append(GINI_Index_1)

p_wt = np.array(weights)
GINI_Index= np.array(GINI_Index)

    
p_wt,GINI_Index
```
