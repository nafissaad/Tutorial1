```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
scaled=pd.read_excel('NEDL_Portfolio_Evaluation.xlsx',index_col='Unnamed: 0')[['Artisan Global Opportunities','Barings Fixed Income','S&P 500','US 10-year bond yield']].dropna()
scaled
```

```python
scaled['Monthly risk-free']=(1+scaled['US 10-year bond yield']/100)**(1/12)-1
scaled['Return_A']=(scaled['Artisan Global Opportunities']/scaled['Artisan Global Opportunities'].shift(1))-1
scaled['Return_B']=(scaled['Barings Fixed Income']/scaled['Barings Fixed Income'].shift(1))-1
scaled['Return_C']=(scaled['S&P 500']/scaled['S&P 500'].shift(1))-1
scaled['Peak_A']=scaled['Artisan Global Opportunities'].expanding().max()
scaled['Peak_B']=scaled['Barings Fixed Income'].expanding().max()
scaled['Peak_C']=scaled['S&P 500'].expanding().max()
scaled['Drawdown_A']=(scaled['Artisan Global Opportunities'].values/scaled['Peak_A'].values)-1
scaled['Drawdown_B']=(scaled['Barings Fixed Income'].values/scaled['Peak_B'].values)-1
scaled['Drawdown_C']=(scaled['S&P 500'].values/scaled['Peak_C'].values)-1
scaled
```

```python
scaled['Excess Return_A']=scaled['Return_A'].values-scaled['Monthly risk-free'].values
scaled['Excess Return_B']=scaled['Return_B'].values-scaled['Monthly risk-free'].values
scaled['Excess Return_C']=scaled['Return_C'].values-scaled['Monthly risk-free'].values
scaled
```

```python
scaled['Downside_A']=[i-j if i<j else 0 for i,j in scaled[['Return_A','Monthly risk-free']].values]
scaled['Downside_B']=[i-j if i<j else 0 for i,j in scaled[['Return_B','Monthly risk-free']].values]
scaled['Downside_C']=[i-j if i<j else 0 for i,j in scaled[['Return_C','Monthly risk-free']].values]
scaled
```

```python
matrix=pd.DataFrame(columns=['Artisan Global Opportunities','Barings Fixed Income','S&P 500'],index=['Annualised return','Annualised risk','Annualised risk-free','Sharp ratio','Beta','Treynor ratio',"Alpha(Jensen's)","Maximum drawdown","Average drawdown","Calmer ratio","Sterling ratio","downsides","Semivariance","Semideviation","Sortino ratio"])
matrix
```

```python
matrix.loc['Annualised return','Artisan Global Opportunities']=(1+scaled['Return_A']).prod()**(1/12)-1
matrix
```

```python
matrix.loc['Annualised return','Artisan Global Opportunities']=(1+scaled['Return_A']).prod()**(1/12)-1
matrix.loc['Annualised return','Barings Fixed Income']=(1+scaled['Return_B']).prod()**(1/12)-1
matrix.loc['Annualised return','S&P 500']=(1+scaled['Return_C']).prod()**(1/12)-1
matrix.loc['Annualised risk','Artisan Global Opportunities']=np.sqrt(12)*scaled['Return_A'].std()
matrix.loc['Annualised risk','Barings Fixed Income']=np.sqrt(12)*scaled['Return_B'].std()
matrix.loc['Annualised risk','S&P 500']=np.sqrt(12)*scaled['Return_C'].std()
matrix.loc['Annualised risk-free','Artisan Global Opportunities']=0.0402
matrix.loc['Annualised risk-free','Barings Fixed Income']=0.0402
matrix.loc['Annualised risk-free','S&P 500']=0.0402
matrix.loc['Sharp ratio','Artisan Global Opportunities']=(matrix.loc['Annualised return','Artisan Global Opportunities']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Annualised risk','Artisan Global Opportunities']
matrix.loc['Sharp ratio','Barings Fixed Income']=(matrix.loc['Annualised return','Barings Fixed Income']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Annualised risk','Barings Fixed Income']
matrix.loc['Sharp ratio','S&P 500']=(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Annualised risk','S&P 500']
matrix.loc['Beta','Artisan Global Opportunities']=np.polyfit(scaled['Return_C'].dropna().values,scaled['Return_A'].dropna().values,1)[0]
matrix.loc['Beta','Barings Fixed Income']=np.polyfit(scaled['Return_C'].dropna().values,scaled['Return_B'].dropna().values,1)[0]
matrix.loc['Beta','S&P 500']=np.polyfit(scaled['Return_C'].dropna().values,scaled['Return_C'].dropna().values,1)[0]
matrix.loc['Treynor ratio','Artisan Global Opportunities']=(matrix.loc['Annualised return','Artisan Global Opportunities']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Beta','Artisan Global Opportunities']
matrix.loc['Treynor ratio','Barings Fixed Income']=(matrix.loc['Annualised return','Barings Fixed Income']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Beta','Barings Fixed Income']
matrix.loc['Treynor ratio','S&P 500']=(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Beta','S&P 500']
matrix.loc["Alpha(Jensen's)",'Artisan Global Opportunities']=(matrix.loc['Annualised return','Artisan Global Opportunities']-matrix.loc['Annualised risk-free','S&P 500'])-matrix.loc['Beta','Artisan Global Opportunities']*(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','Artisan Global Opportunities'])
matrix.loc["Alpha(Jensen's)",'Barings Fixed Income']=(matrix.loc['Annualised return','Barings Fixed Income']-matrix.loc['Annualised risk-free','S&P 500'])-matrix.loc['Beta','Barings Fixed Income']*(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','Barings Fixed Income'])
matrix.loc["Alpha(Jensen's)",'S&P 500']=(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','S&P 500'])-matrix.loc['Beta','S&P 500']*(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','S&P 500'])
matrix
```

```python
matrix.loc['Maximum drawdown','Artisan Global Opportunities']=-np.min(scaled['Drawdown_A'])
matrix.loc['Maximum drawdown','Barings Fixed Income']=-np.min(scaled['Drawdown_B'])
matrix.loc['Maximum drawdown','S&P 500']=-np.min(scaled['Drawdown_C'])
matrix.loc['Average drawdown','Artisan Global Opportunities']=-np.mean(scaled['Drawdown_A'][scaled['Drawdown_A']<0])
matrix.loc['Average drawdown','Barings Fixed Income']=-np.mean(scaled['Drawdown_B'][scaled['Drawdown_B']<0])
matrix.loc['Average drawdown','S&P 500']=-np.mean(scaled['Drawdown_C'][scaled['Drawdown_C']<0])
matrix.loc['Calmer ratio','Artisan Global Opportunities']=(matrix.loc['Annualised return','Artisan Global Opportunities']-matrix.loc['Annualised risk-free','Artisan Global Opportunities'])/matrix.loc['Maximum drawdown','Artisan Global Opportunities']
matrix.loc['Calmer ratio','Barings Fixed Income']=(matrix.loc['Annualised return','Barings Fixed Income']-matrix.loc['Annualised risk-free','Barings Fixed Income'])/matrix.loc['Maximum drawdown','Barings Fixed Income']
matrix.loc['Calmer ratio','S&P 500']=(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Maximum drawdown','S&P 500']
matrix.loc['Sterling ratio','Artisan Global Opportunities']=(matrix.loc['Annualised return','Artisan Global Opportunities']-matrix.loc['Annualised risk-free','Artisan Global Opportunities'])/matrix.loc['Average drawdown','Artisan Global Opportunities']
matrix.loc['Sterling ratio','Barings Fixed Income']=(matrix.loc['Annualised return','Barings Fixed Income']-matrix.loc['Annualised risk-free','Barings Fixed Income'])/matrix.loc['Average drawdown','Barings Fixed Income']
matrix.loc['Sterling ratio','S&P 500']=(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Average drawdown','S&P 500']
matrix.loc['downsides','Artisan Global Opportunities']=scaled[scaled['Downside_A']!=0]['Downside_A'].count()
matrix.loc['downsides','Barings Fixed Income']=scaled[scaled['Downside_B']!=0]['Downside_B'].count()
matrix.loc['downsides','S&P 500']=scaled[scaled['Downside_C']!=0]['Downside_C'].count()
matrix.loc['Semivariance','Artisan Global Opportunities']=np.sum((scaled['Downside_A'])**2)/matrix.loc['downsides','Artisan Global Opportunities']
matrix.loc['Semivariance','Barings Fixed Income']=np.sum((scaled['Downside_B'])**2)/matrix.loc['downsides','Barings Fixed Income']
matrix.loc['Semivariance','S&P 500']=np.sum((scaled['Downside_C'])**2)/matrix.loc['downsides','S&P 500']
matrix.loc['Semideviation','Artisan Global Opportunities']=np.sqrt(matrix.loc['Semivariance','Artisan Global Opportunities'])*np.sqrt(12)
matrix.loc['Semideviation','Barings Fixed Income']=np.sqrt(matrix.loc['Semivariance','Barings Fixed Income'])*np.sqrt(12)
matrix.loc['Semideviation','S&P 500']=np.sqrt(matrix.loc['Semivariance','S&P 500'])*np.sqrt(12)
matrix.loc['Sortino ratio','Artisan Global Opportunities']=(matrix.loc['Annualised return','Artisan Global Opportunities']-matrix.loc['Annualised risk-free','Artisan Global Opportunities'])/matrix.loc['Semideviation','Artisan Global Opportunities']
matrix.loc['Sortino ratio','Barings Fixed Income']=(matrix.loc['Annualised return','Barings Fixed Income']-matrix.loc['Annualised risk-free','Barings Fixed Income'])/matrix.loc['Semideviation','Barings Fixed Income']
matrix.loc['Sortino ratio','S&P 500']=(matrix.loc['Annualised return','S&P 500']-matrix.loc['Annualised risk-free','S&P 500'])/matrix.loc['Semideviation','S&P 500']
matrix
```
