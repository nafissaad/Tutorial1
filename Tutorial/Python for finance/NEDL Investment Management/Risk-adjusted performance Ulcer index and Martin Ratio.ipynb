```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
scaled=pd.read_excel('C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\Technical analysis\\NEDL_Ulcer_Martin.xlsx',index_col='Date')[0:]
scaled=scaled.rename(columns=scaled.iloc[0])
scaled=scaled[1:]
scaled=scaled.iloc[:,0:2]
scaled
```

```python
returns=scaled.pct_change()
returns=returns.fillna(0)
returns
```

```python
returns['Drawdown_A']=np.nan
returns['Drawdown_B']=np.nan
returns.loc['2016-12-30',['Drawdown_A','Drawdown_B']]=0
returns
```

```python
x=[0]
for i in range(1,len(returns)):
    x.append(min(((1+x[-1])*(1+returns.iloc[i]['Blackrock'])-1),0,1))
returns['Drawdown_A']=x
returns
```

```python
x=[0]
for i in range(1,len(returns)):
    x.append(min(((1+x[-1])*(1+returns.iloc[i]['S&P 500'])-1),0,1))
returns['Drawdown_B']=x
returns
```

```python
returns
```

```python
Matrix=pd.DataFrame(columns=[['Blackrock','S&P 500']],index=['Risk-free','Return','Maximum drawdown','Average Drawdown','Ulcer index','Calmer ratio','Sterling ratio','Martin ratio'])
Matrix
```

```python
Matrix.loc['Average Drawdown','Blackrock'][0]
```

```python
Matrix.loc['Risk-free','Blackrock']=0.0165
Matrix.loc['Risk-free','S&P 500']=0.0165
Matrix.loc['Return','Blackrock']=(scaled['Blackrock'][-1]/scaled['Blackrock'][0])**(1/5)-1
Matrix.loc['Return','S&P 500']=(scaled['S&P 500'][-1]/scaled['S&P 500'][0])**(1/5)-1
Matrix.loc['Maximum drawdown','Blackrock']=-min(returns['Drawdown_A'][1:])
Matrix.loc['Maximum drawdown','S&P 500']=-min(returns['Drawdown_B'][1:])
Matrix.loc['Average Drawdown','Blackrock']=-np.mean(returns['Drawdown_A'][1:])
Matrix.loc['Average Drawdown','S&P 500']=-np.mean(returns['Drawdown_B'][1:])
Matrix.loc['Ulcer index','Blackrock']=np.sqrt(np.mean(returns['Drawdown_A'][1:]**2))
Matrix.loc['Ulcer index','S&P 500']=np.sqrt(np.mean(returns['Drawdown_B'][1:]**2))
Matrix.loc['Calmer ratio','Blackrock']=(Matrix.loc['Return','Blackrock'][0]-0.0165)/Matrix.loc['Maximum drawdown','Blackrock'][0]
Matrix.loc['Calmer ratio','S&P 500']=(Matrix.loc['Return','S&P 500'][0]-0.0165)/Matrix.loc['Maximum drawdown','S&P 500'][0]
Matrix.loc['Sterling ratio','Blackrock']=(Matrix.loc['Return','Blackrock'][0]-0.0165)/Matrix.loc['Average Drawdown','Blackrock'][0]
Matrix.loc['Sterling ratio','S&P 500']=(Matrix.loc['Return','S&P 500'][0]-0.0165)/Matrix.loc['Average Drawdown','S&P 500'][0]
Matrix.loc['Martin ratio','Blackrock']=(Matrix.loc['Return','Blackrock'][0]-0.0165)/Matrix.loc['Ulcer index','Blackrock'][0]
Matrix.loc['Martin ratio','S&P 500']=(Matrix.loc['Return','S&P 500'][0]-0.0165)/Matrix.loc['Ulcer index','S&P 500'][0]
Matrix
```

```python
#Black rock Outperformed in Calmer ratio(Calmer ratio Subject to Maximum Draw Down)
#S&P 500 Outperformed in Sterling ratio ratio(Sterling ratio Subject to Average Draw Down)
#S&P 500 Outperformed in martin ratio(Martin Ratio Subject to Ulcer index)
```
