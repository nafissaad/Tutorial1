```python
df=pd.read_excel('C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\The Treynor-Black model-the balance between active and passive investment\\Book3.xlsx',index_col='Unnamed: 0')
df
```

```python
df.info()
```

```python
matrix=pd.DataFrame(columns=[['RDS','BP']],index=['Market price','Fair value','Moth to coverage','Alpha','Beta','Idiosyncratic variance','Apprisal ratio','Active weight','Active weight squared'])
```

```python
matrix.loc['Market price','RDS']=21.53
matrix.loc['Market price','BP']=4.63
matrix.loc['Fair value','RDS']=34.153931
matrix.loc['Fair value','BP']=5.638621
matrix.loc['Moth to coverage','RDS']=36#3 years
matrix.loc['Moth to coverage','BP']=36
matrix.loc['Alpha','RDS']=(matrix.loc['Fair value','RDS']/matrix.loc['Market price','RDS'])**(1/36)-1#suppose to appriciate 1.28% per month
matrix.loc['Alpha','BP']=(matrix.loc['Fair value','BP']/matrix.loc['Market price','BP'])**(1/36)-1
matrix
```

```python
returns=df.pct_change().dropna()
returns
```

```python
np.polyfit(np.array(returns['FTSE 250']),np.array(returns['RDS']),1)#1 means constant true,FTSE 250 is independent 
```

```python
matrix.loc['Beta','RDS']=np.polyfit(np.array(returns['FTSE 250']),np.array(returns['RDS']),1)[0]
matrix.loc['Beta','BP']=np.polyfit(np.array(returns['FTSE 250']),np.array(returns['BP']),1)[0]
matrix
```

```python
returns['RDS Abnormal Returns']=returns['RDS'].values-matrix.loc['Beta','RDS'].values*returns['FTSE 250'].values
returns['BP Abnormal Returns']=returns['BP'].values-matrix.loc['Beta','BP'].values*returns['FTSE 250'].values
returns
```

```python
matrix.loc['Apprisal ratio'].sum()
```

```python
matrix.loc['Idiosyncratic variance','RDS']=returns['RDS Abnormal Returns'].var()
matrix.loc['Idiosyncratic variance','BP']=returns['BP Abnormal Returns'].var()
matrix.loc['Apprisal ratio','RDS']=matrix.loc['Alpha','RDS']/matrix.loc['Idiosyncratic variance','RDS']#Apprisal ratio is risk adjusted return, higher ratio is better
matrix.loc['Apprisal ratio','BP']=matrix.loc['Alpha','BP']/matrix.loc['Idiosyncratic variance','BP']
matrix.loc['Active weight','RDS']=matrix.loc['Apprisal ratio','RDS']/matrix.loc['Apprisal ratio'].sum()
matrix.loc['Active weight','BP']=matrix.loc['Apprisal ratio','BP']/matrix.loc['Apprisal ratio'].sum()
matrix.loc['Active weight squared','RDS']=matrix.loc['Active weight','RDS']**2
matrix.loc['Active weight squared','BP']=matrix.loc['Active weight','BP']**2
matrix
```

```python
Active_alpha=matrix.loc['Active weight'].dot(matrix.loc['Alpha'])#% value
Active_beta=matrix.loc['Active weight'].dot(matrix.loc['Beta'])#normal values
Active_volatility=matrix.loc['Idiosyncratic variance'].dot(matrix.loc['Active weight squared'])#normal value
Market_return=np.prod(1+returns['FTSE 250'])**(1/60)-1#Average monthly Appriciation
Market_variance=returns['FTSE 250'].var()#Marke_variance is less than Idiosyncratic variance and Active_volatiloty which means passive investment is much more attractive 
Market_variance
```

```python
Risk_free_treasury_Annual_rate=0.0079
Risk_free_treasury_monthly=(1+Risk_free_treasury_Annual_rate)**(1/12)-1
Excess_market_return=Market_return-Risk_free_treasury_monthly
Excess_market_return
```

```python
Uncorrelated_active_weight=(Active_alpha/Active_volatility)/(Excess_market_return/Market_variance)#% values
Corrected_active_weight=Uncorrelated_active_weight/(1+(1-Active_beta)*Uncorrelated_active_weight)#% value
passive_weight=1-Corrected_active_weight
Uncorrelated_active_weight,Corrected_active_weight,passive_weight
```

```python
final_weight=pd.DataFrame()
final_weight['RSD']=Corrected_active_weight*matrix.loc['Active weight','RDS'].values
final_weight['BP']=Corrected_active_weight*matrix.loc['Active weight','BP'].values
final_weight['FTSE 250']=passive_weight
final_weight
```
