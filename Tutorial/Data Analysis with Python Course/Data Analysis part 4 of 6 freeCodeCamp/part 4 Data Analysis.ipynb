```python
# Analyzing Data from dat frames
#Question : What is the total number of reported cases and deaths related to Covid-19 in italy?
total_cases=df.new_cases.sum()
total_deaths=df.new_deaths.sum()
```

```python
print('The number of reported cases i s {} and the number of reported deaths is {}.'.format(int(total_cases), int(total_deaths)))
```

```python
#Q What is the overall death rate(ratio of deats to reported cases )?
death_rate=df.new_deaths.sum()/df.new_cases.sum()
```

```python
print('The overall reported death rate is Italy is {:.2f} %.'.format(death_rate*100))
```

```python
#what is the overall number of tests conducted? A total of 935310 tests were conducted
#before daily test numbers were being reported
initial_tests=935310
total_tests=initial_tests+df.new_tests.sum()
```

```python
total_tests
```

```python
#Question What fraction of the test returned a positive result ?
positive_rate=total_cases/total_tests
```

```python
print('{:.2f}% of tests in Italy led to a positive diagnosis.'.format(positive_rate*100))
```
