```python
from bdshare import get_basic_hist_data
import matplotlib.pyplot as plt
import pandas as pd
from pandas.tseries.offsets import MonthEnd
```

```python
tickers='https://raw.githubusercontent.com/nafissaad/Tutorial1/main/Tutorial/My%20projects/BD%20Shares/BD%20Stocks/BD_stocks.csv'
tickers=pd.read_csv(tickers)
tickers
```

```python
tickers=tickers['Trading code']
tickers
```

```python
price,symbols=[],[]
end='2023-02-01'
for symbol in tickers:
    print(symbol)
    df=get_basic_hist_data('2020-01-01',end,symbol,index='date')['close']
    df.index=pd.to_datetime(df.index)
    if not df.empty:
        price.append(df)
        symbols.append(symbol)
    print('End')
```

```python
price
```

```python
all_prices=pd.concat(price,axis=1)
```

```python
all_prices
```

```python
all_prices.columns=symbols
all_prices
```

```python
all_prices.pct_change()
```

```python
(all_prices.pct_change()+1).prod()-1#Cumulative return
```

```python
all_mtl_ret=all_prices.pct_change().resample('M').agg(lambda x:(x+1).prod()-1)#Monthly cumulative return 
all_mtl_ret
```

```python
all_mtl_ret_12=all_mtl_ret.rolling(12).agg(lambda x: (x+1).prod()-1)
all_mtl_ret_12
```

```python
all_mtl_ret_12.dropna(inplace=True)
all_mtl_ret_12
```

```python
curr_=all_mtl_ret_12.iloc[-3]
curr_
```

```python
win_=curr_.nlargest(10)
win_
```

```python
win_ret=all_mtl_ret.loc[win_.name+MonthEnd(1),win_.index]
win_ret
```

```python
win_ret.mean()*100
```

```python
def mom(all_mtl_ret,all_mtl_ret_12):
    
    ret=[]
    
    for i in range(len(all_mtl_ret_12)-1):
        curr_=all_mtl_ret_12.iloc[i]
        win_=curr_.nlargest(10)
        win_ret=all_mtl_ret.loc[win_.name+MonthEnd(1),win_.index]
        ret.append(win_ret.mean())
    return(ret)

    
```

```python
#Random weights for my Choosed stoke 
weights = np.random.random(len(win_.index))
weights /= np.sum(weights)
print('weights :',weights)
print('Total Weights :', np.sum(weights))
```

```python
np.sum(weights * all_prices.pct_change()[win_.index].mean())*252
```

```python
p_ret = []#Return list of our random portfolio
p_vol = []#Volatility list of our random portfolio
p_sr = []#sharp ration to maximize profit and minimize risk(rate of return- risk free rate)/standard deviation
p_wt = []#weights 
for x in range(30000):
    p_weights=np.random.random(len(win_.index))
    p_weights /= np.sum(p_weights)# randomly creating weights
    ret_1=np.sum(p_weights * all_prices.pct_change()[win_.index].mean())*252
    p_ret.append(ret_1)
    
    vol_1 = np.sqrt(np.dot(p_weights.T, np.dot(all_prices.pct_change()[win_.index].cov()*252,p_weights)))
    p_vol.append(vol_1)
    SR_1 = (ret_1-0.03)/vol_1
    p_sr.append(SR_1)
    p_wt.append(p_weights)
p_ret = np.array(p_ret)
p_vol = np.array(p_vol)
p_SR = np.array(p_sr)
p_wt = np.array(p_wt)

    
p_ret,p_vol,p_SR,p_wt
```

```python
np.sum(weights* all_prices.pct_change()[win_.index].dropna().mean())*252
```

```python
SR_idx=np.argmax(p_SR)
wt=p_wt[SR_idx]
wt
```

```python
#Plotting the efficient frontier
ports = pd.DataFrame({'Returns' : p_ret, 'volatility' :p_vol})
ports.plot(x='volatility', y ='Returns', kind='scatter',figsize=(16,9))
plt.scatter(p_vol[SR_idx],p_ret[SR_idx],color="Red",marker='+')
#plt.scatter(x=0.294,y=1.1,color="Red",marker='+')
plt.show()
```

```python
i = 0
while i < len(win_.index):
    print('Stock : %s : %2.2f'%(win_.index[i],(p_wt[SR_idx][i]*100)))
    i += 1
print('\nVolatility :',p_vol[SR_idx])
print('Return :',p_ret[SR_idx])
```
