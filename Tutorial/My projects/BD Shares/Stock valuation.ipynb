```python
#impoer the liberies
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
from bdshare import get_basic_hist_data
plt.style .use('fivethirtyeight')
```

```python
ticker='Olympic'
df=get_basic_hist_data('2019-01-1','2022-10-05',ticker,index='date')
df =df[~(df.open<=0)]
df
```

```python
m=pd.to_datetime(df.index)
m
```

```python
df.set_index(m, inplace=True)
t=df.asfreq('D')
```

```python
#create figure
plt.figure(figsize=(15,10))
df1=df
#define width of candlestick elements
width = 1.3
width2 = 0.2

#define up and down prices
up = df1[df1.close>=df1.open]
down = df1[df1.close<df1.open]

#define colors to use
col1 = 'green'
col2 = 'red'

#plot up prices
plt.bar(up.index,up.close-up.open,width,bottom=up.open,color=col1)
plt.bar(up.index,up.high-up.close,width2,bottom=up.close,color=col1)
plt.bar(up.index,up.low-up.open,width2,bottom=up.open,color=col1)

#plot down prices
plt.bar(down.index,down.close-down.open,width,bottom=down.open,color=col2)
plt.bar(down.index,down.high-down.close,width2,bottom=down.close,color=col2)
plt.bar(down.index,down.low-down.open,width2,bottom=down.open,color=col2)

#rotate x-axis tick labels
plt.xticks(rotation=45, ha='right')

#display candlestick chart
plt.title('Olympic Chart')
plt.xlabel('Dates')
plt.ylabel('Price BDT')
plt.axhline(170, color='k', linestyle='--', alpha=0.2,label='Valuation')
plt.legend( fontsize=16)
plt.show()
```

```python
ticker='RUNNERAUTO'
df=get_basic_hist_data('2019-01-1','2022-10-05',ticker,index='date')
df=df[~(df.open<=0)]
df['adj']=(df.open+df.close)/2

m=pd.to_datetime(df.index)
m
df.set_index(m, inplace=True)
t=df.asfreq('D')
```

```python
df[['open','close','adj']][20:30].plot()
```

```python
df=df.interpolate(method='time')
df
```

```python
stl=STL(df['adj'],period=90)#two month seasonal Decomposition
Result = stl.fit()
```

```python
seasonal,trend,resid = Result.seasonal,Result.trend, Result.resid
df['trend']=trend
df
```

```python
#create figure
plt.figure(figsize=(15,10))
df1=df#[-100:]
#define width of candlestick elements
width = 1.3
width2 = 0.2

#define up and down prices
up = df1[df1.close>=df1.open]
down = df1[df1.close<df1.open]

#define colors to use
col1 = 'green'
col2 = 'red'

#plot up prices
plt.bar(up.index,up.close-up.open,width,bottom=up.open,color=col1)
plt.bar(up.index,up.high-up.close,width2,bottom=up.close,color=col1)
plt.bar(up.index,up.low-up.open,width2,bottom=up.open,color=col1)

#plot down prices
plt.bar(down.index,down.close-down.open,width,bottom=down.open,color=col2)
plt.bar(down.index,down.high-down.close,width2,bottom=down.close,color=col2)
plt.bar(down.index,down.low-down.open,width2,bottom=down.open,color=col2)

#
plt.plot(np.array(df1.index),df1.trend.values,label='trend')

#rotate x-axis tick labels
plt.xticks(rotation=45, ha='right')

#display candlestick chart
plt.title('Runner Auto')
plt.xlabel('Dates')
plt.ylabel('Price BDT')
plt.axhline(70, color='k', linestyle='--', alpha=0.2,label='Valuation')
plt.legend( fontsize=16)
plt.show()
```
