```python
df5=df['In which year your family had come to this place/area?'].apply(lambda x: pd.to_datetime(str(int(x))))
df5
```

```python
dummies=pd.get_dummies(df['Area of the Survey'])
dummies
```

```python
(df7.index.year//10)*10
```

```python
df8=df7.groupby((df7.index.year//10)*10).sum()[1:]
df8
```

```python
#plt.figure(figsize=(10, 10), dpi=200)
df8.plot.bar(stacked=True,figsize=(12, 8))
plt.xlabel('Year',fontsize=15)
plt.ylabel('Number of family',fontsize=15)
plt.title('Number of family Migrated to this place',fontsize=15)
plt.xticks(rotation=45)
plt.savefig('C:\\Users\\nafissaad\\Tutorial\\Assignment\\1.jpg',bbox_inches='tight',pad_nches=2,transperent=True)
plt.show()
```

```python
x=df.groupby('Reason for migration')['Reason for migration'].agg('count').sort_values(ascending=False)
x
```

```python
len(x[x<=2])
```

```python
Reason_for_migration_3=x[x<=2]
Reason_for_migration_3
```

```python
df['Reason for migration']=df['Reason for migration'].apply(lambda x: 'Other' if x in Reason_for_migration_3 else x)
k=pd.DataFrame(df.groupby('Reason for migration')['Reason for migration'].agg('count'))
```

```python
dummies1=df['Reason for migration']
dummies2=pd.get_dummies(df['Area of the Survey'])
k=pd.concat([dummies1,dummies2],axis='columns')
k
```

```python
mmm=k.groupby(['Reason for migration'])[['Bondor Tila','Jahajmara','Namarpara']].sum()
mmm
```

```python
mmm.plot.bar(rot=0,figsize=(20, 12),width=0.9,color=['c',(0.2, 0.4, 0.6, 0.6),'orange'])
plt.xlabel('Year',fontsize=15)
plt.ylabel('Number of family',fontsize=15)
plt.title('Number of family Migrated to this place',fontsize=15)
plt.xticks(rotation=90)
plt.savefig('C:\\Users\\nafissaad\\Tutorial\\Assignment\\4.jpg',bbox_inches='tight',pad_nches=2,transperent=True)
plt.show()
```
