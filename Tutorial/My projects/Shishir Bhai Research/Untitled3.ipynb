```python
dummies=pd.get_dummies(df['What was your "JSC" /JDC result?'])
dummies=pd.concat([df['Your private Tuition fee amount'],dummies],axis='columns')
dummies=dummies.dropna()
dummies
```

```python
def convert_sqft_to_num(x):
    if type(x)== str:
        tokens = x.split(' ')
        if len(tokens)==2:
            return(0)
        try:
            return float(x)
        except:
            return None
    else:
        return(x)
    
convert_sqft_to_num('No tuition fees')
```

```python
dummies=dummies.dropna()
dummies['Your private Tuition fee amount']=dummies['Your private Tuition fee amount'].apply(convert_sqft_to_num)
plt.figure(figsize=(16,5))
plt.scatter(dummies['Your private Tuition fee amount'],dummies['A+'],marker='+',color="red")
plt.title('Private Tuition Fee Amount')
plt.xlabel('Tuition Fee')
plt.ylabel('Probability of Getting A+')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(dummies[['Your private Tuition fee amount']],dummies['A+'],train_size=0.90)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train.values, y_train.values)
```

```python
model.score(X_test.values,y_test.values)
```

```python
m=model.predict(dummies[['Your private Tuition fee amount']].values)
```
