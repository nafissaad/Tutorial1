```python
Balance_Sheet.columns=[str(i) for i in Balance_Sheet.columns]
Balance_Sheet.columns=pd.to_datetime(Balance_Sheet.columns)+pd.offsets.YearEnd()#pd.offsets.MonthEnd
Balance_Sheet

Income_Satetment.columns=[str(i) for i in Income_Satetment.columns]
Income_Satetment.columns=pd.to_datetime(Income_Satetment.columns)+pd.offsets.YearEnd()#pd.offsets.MonthEnd
Income_Satetment
```

```python
Balance_Sheet=Balance_Sheet.dropna(axis=0)
Income_Satetment=Income_Satetment.dropna(axis=0)
Income_Satetment
```

```python
Balance_Sheet.index
```

```python
Total_Current_Assets=Balance_Sheet.loc['    Total Current Assets',Balance_Sheet.columns].values
Total_Current_Liabilities=Balance_Sheet.loc['    Total Current Liabilities',Balance_Sheet.columns].values
Cash_and_Cash_Equivalents=Balance_Sheet.loc['            Cash and Cash Equivalents',Balance_Sheet.columns].values
Net_sales=Income_Satetment.loc['    Total Revenue',Income_Satetment.columns].values#net sales=revenue or total revenue
Total_Assets=Balance_Sheet.loc['Total Assets',Balance_Sheet.columns].values
Total_Accounts_Receivable=Balance_Sheet.loc['        Trade and Other Receivables, Current',Balance_Sheet.columns].values
Total_Liabilities=Balance_Sheet.loc['Total Liabilities',Balance_Sheet.columns].values
total_equity=Balance_Sheet.loc['Total Equity',Balance_Sheet.columns].values
net_income=Income_Satetment.loc['Net Income Available to Common Stockholders',Income_Satetment.columns].values
```

```python
current_ratio=Total_Current_Assets/Total_Current_Liabilities#1.2 to 2 is considered as ideal current ration,which means the company has 2 time asset than its liabelities
current_ratio
```

```python
cash_ratio=Cash_and_Cash_Equivalents/Total_Current_Liabilities#for every tk the company owes creditors in the next 12 months it has 0.5 in cash. 1 is considered to be the ideal cash ratio.
cash_ratio
```

```python
Total_Asset_turnover=Net_sales/Total_Assets
Total_Asset_turnover#each tk of assets generates 0.51 poisha in sales. The higher the ratio, the more efficient the company is using its assets to make sales.
```

```python
Account_receivable_turnover=Net_sales/Total_Accounts_Receivable
Account_receivable_turnover#A high-efficiency ratio means that the company has high-quality customers who pay their debts in due time.
```

```python
Debt_Ratio=Total_Liabilities/Total_Assets
Debt_Ratio#Debt financing
```

```python
Equity_Ratio=total_equity/Total_Assets#Equity financed by owner. 0.5 tk means half of the firm financed by the owners and other half by creditors 
Equity_Ratio
```

```python
Profit_margin_ratio=net_income/Net_sales#availability to generate net income, 10% means avg ,20 means high ,5% mean poor
Profit_margin_ratio
```

```python
ROE=net_income/total_equity
ROE
```

```python
ROA=net_income/Total_Assets
ROA
```
