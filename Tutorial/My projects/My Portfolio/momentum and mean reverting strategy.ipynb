```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from bdshare import get_basic_hist_data
import yfinance as yf
```

```python
df=pd.read_excel("C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\My Portfolio\\data_set.xlsx",index_col='date')
df
```

```python
x=0
idx=[]
for i in df.index:
    if x<5:
        x=x+1
    elif x==5:
        x=1
        idx.append(i)
```

```python
df1=df.loc[idx]
df1
```

```python
df1.index=pd.DataFrame(df1.index)
```

```python
pct_change=pd.DataFrame()
```

```python
pct_change['Yearly']=df1['UNIQUEHRL'].pct_change(52)
pct_change['Weekly']=df1['UNIQUEHRL'].pct_change()
pct_change
```

```python
pct_change=pct_change.dropna()
pct_change
```

```python
((pct_change['Yearly']+(pct_change['Weekly'].shift(1)*-1))*pct_change['Weekly']).cumsum().plot(label='X')
pct_change['Weekly'].cumsum().plot()
plt.legend()
```

```python
(((pct_change['Yearly']+(pct_change['Weekly'].shift(1)*-1))*10000)*pct_change['Weekly']).cumsum().plot()
(pct_change['Weekly']*10000).cumsum().plot()
```
