```python
#Calculate the MACF and single line indicators
#Claculate the short term exponential moving average(EMA)
ShortEMA=df.UNIQUEHRL.ewm(span=12,adjust=False).mean()
#calculate the long term exponential moving average (EMA)
LongEMA=df.UNIQUEHRL.ewm(span=26,adjust=False).mean()
#Calculate the MACD line
MACD=ShortEMA-LongEMA
#Calculate The Signal line
Signal=MACD.ewm(span=9,adjust=False).mean()
```

```python
def plot_macd(prices, macd, signal, hist):
    ax1 = plt.subplot2grid((16,1), (0,0), rowspan = 5, colspan = 1)
    ax2 = plt.subplot2grid((16,1), (6,0), rowspan = 3, colspan = 1)

    ax1.plot(prices)
    ax2.plot(macd, color = 'grey', linewidth = 1.5, label = 'MACD')
    ax2.plot(signal, color = 'skyblue', linewidth = 1.5, label = 'SIGNAL')

    for i in range(len(prices)):
        if hist[i]>0 and hist[i]>hist[i-1]:
            ax2.bar(prices.index[i], hist[i], color = '#26a69a')
            
        elif hist[i]<hist[i-1]:
            ax2.bar(prices.index[i], hist[i], color = '#ef5350')
            
        elif hist[i]>hist[i-1]:
            ax2.bar(prices.index[i], hist[i], color = '#26a69a')
       
        elif hist[i]<0 and hist[i]<hist[i-1]:
            ax2.bar(prices.index[i], hist[i], color = '#ef5350')
    plt.legend(loc = 'lower left')
```

```python
plt.figure(figsize=(8, 9), dpi=200)
plot_macd(prices = df['UNIQUEHRL'], macd = MACD, signal = Signal, hist =  MACD)
```
