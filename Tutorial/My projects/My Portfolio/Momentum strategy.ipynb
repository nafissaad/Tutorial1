```python
ret_df=df.pct_change()
ret_df
```

```python
ret_df.index=pd.to_datetime(ret_df.index)
ret_df
```

```python
mtl_ret=(ret_df+1).resample('M').prod()
mtl_ret
```

```python
mtl_12=mtl_ret.rolling(12).apply(np.prod).dropna()#yraely ret
mtl_12
```

```python
top_=mtl_12.loc['2022-05-31'].nlargest(20)#top 5 performer on 2022-10-31 from last 12 month
top_
```

```python
def top_performers(date):
    #print(date)
    top_=mtl_12.loc[date].nlargest(20)
    relevant_ret=mtl_ret[top_.name:][1:2][top_.index]
    #print((relevant_ret).mean(axis=1).values[0])
    return((relevant_ret).mean(axis=1).values[0])
```

```python
top_performers('2022-05-31')
```

```python
mom_ret=[]
for date in mtl_12.index[:-1]:
    mom_ret.append(top_performers(date))
```
