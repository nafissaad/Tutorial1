```python
import statsmodels.api as sm
def VIF_score(df):
    VIF_score=[]
    ticker=[]
    for i in df.columns:
        model=sm.OLS(df[i],sm.add_constant(df.loc[:, df.columns!=i]),hasconst=True)
        results=model.fit()
        VIF_score.append(results.rsquared)
        ticker.append(i)
    return(VIF_score,ticker)  
```

```python
def sharpe_pf(W,portfolio_ret):
    pf_risk=(W.T.dot(portfolio_ret.cov().dot(W)))**(1/2)
    SR=W.dot(portfolio_ret.mean())/pf_risk
    return(SR)
```

```python
p_ret = []#Return list of our random portfolio
p_vol = []#Volatility list of our random portfolio
p_sr = []#sharp ration to maximize profit and minimize risk(rate of return- risk free rate)/standard deviation
p_wt = []#weights

for x in range(30000):
    p_weights=np.random.random(len(sets.columns))
    p_weights/=np.sum(p_weights)
    ret_1=p_weights.dot(returns.mean())*252
    p_ret.append(ret_1)
    vol_1 = np.sqrt(p_weights.T.dot((portfolio_ret.cov()*252).dot(p_weights)))
    p_vol.append(vol_1)
    SR_1 = (ret_1-0.06)/vol_1
    p_sr.append(SR_1)
    p_wt.append(p_weights)
p_ret = np.array(p_ret)
p_vol = np.array(p_vol)
p_SR = np.array(p_sr)
p_wt = np.array(p_wt)

    
p_ret,p_vol,p_SR,p_wt
```

```python
def my_port(investment):
    Q=[]
    P=[]
    T=[]
    m=pd.read_excel("C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\My Portfolio\\Weights.xlsx",index_col='Unnamed: 0')
    pr=price(sdate,edate)
    for i in sets.columns:
        wt=(investment/ m['Weights'].sum())*(m.loc[i].values)
        Q.append(round(float(wt)/float(pr[i].values)))
        P.append(float(pr[i].values))
        T.append(round(float(wt)/float(pr[i].values))*float(pr[i].values))
    return(Q,P,T)
        
```
