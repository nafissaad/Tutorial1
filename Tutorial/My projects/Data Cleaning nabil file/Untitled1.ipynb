```python
import pandas as pd
import numpy as np
```

```python
dm=pd.DataFrame()
dm['Company Ticker']=[]
dm['Company Name']=np.nan
dm['Eps Record']=np.nan
dm['Cap']=np.nan
dm['Chart']=np.nan
dm['Valuation Model']=np.nan
dm['CR']=np.nan
dm['A']=np.nan
dm['B']=np.nan
dm['C']=np.nan
```

```python
def word_spliter1(x):
        tokens = x.split('\xa0(\xa0')
        if len(tokens)>=2:
            return(tokens[0])
        try:
            return(x)
        except:
            return None
```

```python
def word_spliter2(x):
        tokens = x.split('\xa0(\xa0')
        if len(tokens)>=2:
            return(tokens[1])
        try:
            return(x)
        except:
            return None
```

```python
Company_Ticker=[]

for i in range(1,10):
    print(i)
    df=pd.read_excel('C:\\Users\\nafissaad\\Downloads\\FFF14100.xlsx','Sheet'+str(i))
    df1=df[df.columns[0]]
    for i in range(len(df1)):
        try:
            x=word_spliter1(df1[i])
            print(x)
            Company_Ticker.append(x)
        except Exception:
            pass
```

```python
Company_Name=[]
for i in range(1,10):
    df=pd.read_excel('C:\\Users\\nafissaad\\Downloads\\FFF14100.xlsx','Sheet'+str(i))
    df1=df[df.columns[0]]
    for i in range(len(df1)):
        try:
            x=word_spliter2(df1[i])
            print(x)
            Company_Name.append('('+x)
        except Exception:
            pass
```

```python
dm['Company Name']=Company_Name
```

```python
dm['Company Ticker']=Company_Ticker
dm
```
