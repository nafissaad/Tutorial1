```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df=pd.read_csv('C:\\Users\\nafissaad\\API_FP.CPI.TOTL_DS2_en_csv_v2_4751116.csv',skiprows=4)
df=df.T
df1=df[0:4]
df1=df1.T
df1
```

```python
bgd=pd.DataFrame()
bgd['Consumer price index (2010 = 100)']=df['Bangladesh']
```

```python
plt.figure(figsize=(15.2,9.5))
plt.plot(np.array(bgd.index),bgd['Consumer price index (2010 = 100)'].values)
plt.xlabel('Year')
plt.ylabel(bgd.columns[0])
plt.title(bgd.columns[0]+ 'line Chart')
plt.xticks(rotation=45)
plt.savefig('C:\\Users\\nafissaad\\Tutorial\\1.png',bbox_inches='tight',pad_nches=2,transperent=True)
plt.show()
```

```python
plt.figure(figsize=(15.2,9.5))
plt.plot(np.array(bgd.index),bgd['Inflation'].values)
plt.xlabel('Year')
plt.ylabel(bgd.columns[1]+' Rate(%)')
plt.title(bgd.columns[1]+ ' line Chart')
plt.xticks(rotation=45)
plt.savefig('C:\\Users\\nafissaad\\Tutorial\\2.png',bbox_inches='tight',pad_nches=2,transperent=True)
plt.show()
```

```python
plt.figure(figsize=(15.2,9.5))
plt.plot(np.array(bgd.index),bgd['Real Interest Rate'].values,label='Real')
plt.plot(np.array(bgd.index),bgd['Nominal Interest Rate'].values,label='Nominal')
plt.xlabel('Year')
plt.ylabel(' Interest Rate (%)')
plt.title('Real And Nominal Interest Rate Line Chart')
plt.xticks(rotation=45)
plt.legend()
plt.savefig('C:\\Users\\nafissaad\\Tutorial\\3.png',bbox_inches='tight',pad_nches=2,transperent=True)
plt.show()
```
