```python
def cleaning(x):
    i=x.split('-')
    x0=i[0].strip('[')
    x1=i[1].strip(')')
    x0=0 if x0=='' else x0
    x1=0 if x1=='' else x0
    return(float(x0)+float(x1))

def cleaning_age(x):
    if x==df['AGE'].value_counts().index[6]:
        x0=x.strip('+')
        x1=0
        
    else:
        i=x.split('-')
        x0=i[0]
        x1=i[1]
        
    return((float(x0)+float(x1))/2)
```

```python
df['INCOME']=df['INCOME'].map(lambda x: cleaning(x))
```

```python
df['AGE']=df['AGE'].map(lambda x: cleaning_age(x))
```
