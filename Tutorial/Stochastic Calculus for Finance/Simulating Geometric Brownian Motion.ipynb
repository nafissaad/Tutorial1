```python
# Simulateting Geometric Brownian Motion(GBM)
#S is stock price,S0 is initial Stock price
#the dynamics as defined in the following Stocastic Differential Equaation(SDE)
```

# \Large S_t = S_0 {\rm e}^{(\mu-\frac{\sigma^2}{2})t + \sigma W_t}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#drift coefficient
mu=0.1
#number of steps
n=100
#time in years
T=1
#number of simulation
M=100
#Initial Stock Price
S0=100
#volatilaity
sigma=0.3
```

```python
#simulate GBM Paths
#calculate each time Step
dt=T/n
#Stimulate stock price at time t Using Numpy array
St=np.exp((mu - sigma ** 2 / 2) * dt+ sigma * np.random.normal(0, np.sqrt(dt), size=(M,n)).T)
#Including inital price of the stock
St=np.vstack([np.ones(M),St])
#Otcomes of the stimulations
St = S0 * St.cumprod(axis=0)
```

```python
#Consider Time Inervals in years
time = np.linspace(0,T,n+1)
#required Numpy array that is same shape as St
tt = np.full(shape=(M,n+1), fill_value=time).T
```

```python
plt.plot(tt,St)
plt.xlabel("Years (t)")
plt.ylabel("Stock Price (St)")
plt.title(
    "Realizations of Geometric Brownian Motion\n dSt=μStdt+σStdWt\n S0=0,μ=1,σ=2".format(S0, mu, sigma)
)
plt.show()
```
