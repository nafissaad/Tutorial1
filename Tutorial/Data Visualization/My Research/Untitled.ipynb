```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
plt.style.use('fivethirtyeight')
```

```python
df=pd.read_excel('C:\\Users\\nafissaad\\Downloads\\2017-3-30-062.xlsx',"Data2015",skiprows=1)
df
```

```python
df.info()
```

```python
df=df[['Per capita GNI rank','HDI rank']]
Test=pd.DataFrame(data={
    'values':[0,25,50,75,100,125,150,175],
})
Test
```

```python
reg= linear_model.LinearRegression()
fit=reg.fit(df[['HDI rank']].values,df['Per capita GNI rank'])
fit.coef_
```

```python
plt.figure(figsize=(12.2,6.4))
plt.scatter(x=df['Per capita GNI rank'],y=df['HDI rank'],color='green')
plt.plot(np.array(Test.values),reg.predict(Test.values),color='red')
plt.xlabel('Per Capita GNI Rank')
plt.ylabel('HDI Rank')
plt.title('Per Capita GNI Rank Vs HDI Rank')
plt.show()
```
