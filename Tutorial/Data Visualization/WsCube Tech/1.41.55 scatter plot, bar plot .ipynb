```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
colors=['r','g','y','b','r','g','r']
size=[100,200,300,400,500,600,700]
plt.scatter(x,y,c=colors,s=size,alpha=0.8,marker='*',edgecolor='g',linewidth=2)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
y2=[3,4,2,5,1,6,2]
colors=[10,20,30,40,50,60,70]
size=[100,200,300,400,500,600,700]
mappable=plt.scatter(np.array(x),y,c=colors,s=size,cmap='viridis',alpha=0.8)
plt.scatter(np.array(x),y2,color='r',s=size,alpha=0.8)
t=plt.colorbar(mappable)
t.set_label('ColorBar')
plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.show()
```

```python
#Bar plot
import matplotlib.pyplot as plt
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel('Language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.title('wscube',fontsize=15)
c=['y','b','m','g']
plt.bar(x,y,width=0.4,color=c,align='center',edgecolor='r',linewidth=5,linestyle=':',alpha=0.4,label='wscube')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel('Language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.title('wscube',fontsize=15)
plt.bar(p,y,width,label='wscube')
plt.bar(p1,z,width,color='y',label='wscube')
plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel('Language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.title('wscube',fontsize=15)
width=0.2
plt.bar(p,y,width=0.4,label='wscube')
plt.bar(p1,z,width=0.4,color='y',label='wscube')
plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel('Language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.title('wscube',fontsize=15)
width=0.2
plt.barh(p,y,width,label='wscube')
plt.barh(p1,z,width,color='y',label='wscube')
plt.yticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```
