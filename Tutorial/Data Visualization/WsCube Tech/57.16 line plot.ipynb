```python
x=['python','c','c++','java']
y=[85,70,60,82]
plt.title('wscube',fontsize=15)
plt.bar(x,y,width=0.4,color=c,edgecolor='r',linewidth=5,linestyle=':',alpha=0.5,label='popularity')
plt.xlabel('language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
plt.title('wscube',fontsize=15)
plt.bar(x,y,width=0.4,color='r',edgecolor='r',linewidth=5,linestyle=':',alpha=0.5,label='popularity')
plt.bar(x,z,width=0.4,color='y',edgecolor='y',linewidth=5,linestyle=':',alpha=0.5,label='popularity2')
plt.xlabel('language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.legend()
plt.show()
```

```python
import numpy as np
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.title('wscube',fontsize=15)
plt.bar(p,y,width,color='r',edgecolor='r',linewidth=5,linestyle=':',alpha=0.5,label='popularity')
plt.bar(p1,z,width,color='y',edgecolor='y',linewidth=5,linestyle=':',alpha=0.5,label='popularity2')
plt.xlabel('language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.xticks(p+width,x)
plt.legend()
plt.show()
```

```python
import numpy as np
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.title('wscube',fontsize=15)
plt.bar(p,y,width,color='r',edgecolor='r',linewidth=5,linestyle=':',alpha=0.5,label='popularity')
plt.bar(p1,z,width,color='y',edgecolor='y',linewidth=5,linestyle=':',alpha=0.5,label='popularity2')
plt.xlabel('language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
import numpy as np
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.title('wscube',fontsize=15)
plt.bar(p,y,width,color='r',edgecolor='r',linewidth=5,linestyle=':',alpha=0.5,label='popularity')
plt.bar(p1,z,width,color='y',edgecolor='y',linewidth=5,linestyle=':',alpha=0.5,label='popularity2')
plt.xlabel('language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.xticks(p+width/2,x,rotation=45)
plt.legend()
plt.show()
```

```python
import numpy as np
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.title('wscube',fontsize=15)
plt.barh(x,y,color='r',edgecolor='r',label='popularity')
plt.barh(x,z,color='y',edgecolor='y',label='popularity2')
plt.ylabel('language',fontsize=15)
plt.xlabel('No.',fontsize=15)
plt.yticks(p+width/2,x,rotation=45)
plt.legend()
plt.show()
```
