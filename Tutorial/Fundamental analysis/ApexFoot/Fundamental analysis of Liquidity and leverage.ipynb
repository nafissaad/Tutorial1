```python
def leverage(Balance_Sheet,Cash_flow,Income_Statement):
    #Score 6 long-term debt ratio
    try:
        lt_debt=Balance_Sheet.loc['Long-Term Debt',Balance_Sheet.columns[4]]
        total_assets=Balance_Sheet.loc['Total Assets',Balance_Sheet.columns[4]]
        debt_ratio=lt_debt/total_assets
        debt_ratio_score=1 if debt_ratio <0.4 else 0
    except:
        debt_ratio_score=1#Means the company doesnt have any kind of long-term debt
        
    #Score 7 - Current ratio
    current_assets=Balance_Sheet.loc['Total Current Assets',Balance_Sheet.columns[4]]
    current_lib=Balance_Sheet.loc['Total Current Liabilities',Balance_Sheet.columns[4]]
    current_ratio=current_assets/current_lib
    current_ratio_score=1 if current_ratio >1 else 0
    leverage_score=debt_ratio_score+current_ratio_score
    print(leverage_score)
```

```python
leverage(Balance_Sheet,Cash_flow,Income_Statement)
```
