```python
def convert_num(x):
    if type(x)==float:
        x=0
    else:
        x=str(x)
        tokens=x.split('%')
        if len(tokens)==1:
            try:
                return(float(tokens[0]))
            except:
                return x
        try:
            return float(float(tokens[0])/100)
        except:
            return x
```

```python
convert_num('aaa')
```
