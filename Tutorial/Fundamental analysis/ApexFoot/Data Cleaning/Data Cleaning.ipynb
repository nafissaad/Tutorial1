```python
def convert_num(x):
    if type(x)==float:
        x=0
    else:
        x=str(x)
        tokens=x.split('%')
        if len(tokens)==1:
            try:
                return(float(tokens[0]))
            except:
                return x
        try:
            return float(float(tokens[0])/100)
        except:
            return x
```

```python
def convert_num1(x):
    try:
        x=x.replace('(','')
        x=x.replace(')','')
        l=x[-1]
        x=x.replace('B','')
        x=x.replace('M','') 
        x=x.replace('K','')
        if l=='K':
            return(float(x))
        elif l=='M':
            return(float(x)*1000)
        elif l=='B':
            return(float(x)*1000000)
        else:
            return(float(x))
    except:
        return(x)
```

```python
#Making a pipline to automate the processing
def Cleaner(path,name):
    Book=pd.read_csv(path+name+'.csv')
    Book=Book.rename(columns = {'Item  Item':'BDT Per Billon'})
    Book=Book.dropna(axis='columns')
    Book=Book.replace('-',np.nan)
    Book.index=Book['BDT Per Billon']
    Book=Book.drop(['BDT Per Billon'],axis='columns')
    Book=Book.dropna(axis='rows',how='all')
    Book=Book.T
    for i in Book.columns:
        Book[i]=Book[i].apply(convert_num).values
    for i in Book.columns:
        Book[i]=Book[i].apply(convert_num1).values
    Book=Book.T
    Book.to_csv('C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\ApexFood\\Data Cleaning\\'+name+'.csv')
```
