```python
def operating_efficiency(Balance_Sheet,Cash_flow,Income_Statement):
    # Secore -8 Gross margin
    gross_profit_margin=(Income_Statement.loc['Sales/Revenue',Income_Statement.columns[4]]-Income_Statement.loc[
        'Cost of Goods Sold (COGS) incl. D&A',Income_Statement.columns[4]])/Income_Statement.loc[
        'Sales/Revenue',Income_Statement.columns[4]]#excluding D&A Depreciation &
    #Amortization Expense Depreciation Amortization of Intangibles Gross Income
    gross_profit_margin_py=(Income_Statement.loc['Sales/Revenue',Income_Statement.columns[3]]-Income_Statement.loc[
        'Cost of Goods Sold (COGS) incl. D&A',Income_Statement.columns[3]])/Income_Statement.loc[
        'Sales/Revenue',Income_Statement.columns[3]]
    gm_score=1 if gross_profit_margin > gross_profit_margin_py else 0
    
    
    #Score 9 - Asset turnover
    Avg_asset=(Balance_Sheet.loc['Total Assets',Balance_Sheet.columns[4]]+
               Balance_Sheet.loc['Total Assets',Balance_Sheet.columns[3]])/2
    Avg_asset_last_year=(Balance_Sheet.loc['Total Assets',Balance_Sheet.columns[3]]+
                         Balance_Sheet.loc['Total Assets',Balance_Sheet.columns[2]])/2
    
    Asset_turnover=Income_Statement.loc['Sales/Revenue',Income_Statement.columns[4]]/Avg_asset
    Asset_turnover_last_year=Income_Statement.loc['Sales/Revenue',Income_Statement.columns[3]]/Avg_asset_last_year
    at_score=1 if Asset_turnover < Asset_turnover_last_year else 0

    
    operating_efficiency_score=gm_score+at_score
    print(operating_efficiency_score)
    return(gross_profit_margin_py,gross_profit_margin,Asset_turnover_last_year,Asset_turnover)
```

```python
operating_efficiency(Balance_Sheet,Cash_flow,Income_Statement)
```
