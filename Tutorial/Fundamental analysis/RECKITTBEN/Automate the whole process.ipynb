```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#Scrapping method 2
def get_data(ticker,x='',i=4):
    return(pd.read_html('https://www.marketwatch.com/investing/stock/'+ticker+'/financials/'+x+'?countrycode=bd')[i])
```

```python
ticker='RECKITTBEN'
Income_statement=get_data(ticker)
Balance_Sheet_A=get_data(ticker,'balance-sheet')
Balance_Sheet_L=get_data(ticker,'balance-sheet',5)
Cash_flow_A=get_data(ticker,'cash-flow')
Cash_flow_B=get_data(ticker,'cash-flow',5)
Cash_flow_C=get_data(ticker,'cash-flow',6)
Balance_Sheet=pd.concat([Balance_Sheet_A,Balance_Sheet_L],axis='rows',)
Cash_flow=pd.concat([Cash_flow_A,Cash_flow_B,Cash_flow_C],axis='rows',)
Balance_Sheet=Balance_Sheet.reset_index()
Balance_Sheet=Balance_Sheet.drop(['index'],axis='columns')
Cash_flow=Cash_flow.reset_index()
Cash_flow=Cash_flow.drop(['index'],axis='columns')
Balance_Sheet
```

```python
def convert_num(x):
    if type(x)==float:
        x=0
    else:
        x=str(x)
        tokens=x.split('%')
        if len(tokens)==1:
            try:
                return(float(tokens[0]))
            except:
                return x
        try:
            return float(float(tokens[0])/100)
        except:
            return x
```

```python
Balance_Sheet.index=Balance_Sheet['BDT Per Billon']
Balance_Sheet=Balance_Sheet.drop(['BDT Per Billon'],axis='columns')
Balance_Sheet
```

```python
def convert_num1(x):
    try:
        t=x[0]
        x=x.replace('(','-')
        x=x.replace(')','')
        x=x.replace(',','')
        l=x[-1]
        x=x.replace('%','')
        x=x.replace('B','')
        x=x.replace('M','') 
        x=x.replace('K','')
        if l=='K':
            return(float(x))*1000
        elif l=='M':
            return(float(x)*1000000)
        elif l=='B':
            return(float(x)*1000000000)
        elif l=='%':
            return(float(x)/100)     
        else:
            return(float(x))
    except:
        return(x)
```

```python
#Making a pipline to automate the processing
def Cleaner(path,name):
    Book=pd.read_csv(path+name+'.csv')
    Book=Book.rename(columns = {'Item  Item':'BDT Per Billon'})
    Book=Book.dropna(axis='columns')
    Book=Book.replace('-',np.nan)
    Book.index=Book['BDT Per Billon']
    Book=Book.drop(['BDT Per Billon'],axis='columns')
    Book=Book.dropna(axis='rows',how='all')
    Book=Book.T
    for i in Book.columns:
        Book[i]=Book[i].apply(convert_num).values
    for i in Book.columns:
        Book[i]=Book[i].apply(convert_num1).values
    Book=Book.T
    Book.to_csv('C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\'+ticker+'\\'+name+'.csv')
```
