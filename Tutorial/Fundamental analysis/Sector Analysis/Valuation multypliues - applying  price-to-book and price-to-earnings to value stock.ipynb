```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_excel('C:\\Users\\nafissaad\\Tutorial\\Fundamental analysis\\NEDL Investment Management\\Book2.xlsx',index_col='Comparables')
df=df[df.columns[0:5]]
df
```

```python
df.info()
```

```python
df.columns[1]
```

```python
df['Book Value of equity']=df['Total assets ']-df['Total liabilities']
df
```

```python
df['Book value per share']=df['Book Value of equity']/df['Shares outstanding']
df
```

```python
df['Price-to-book(P/B)']=df['Market price']/df['Book value per share']
df
```

```python
df['Price-to-earnings(P/E)']=df['Market price']/df['Earnings per share']#how many years it will take o pay its price
df
```

```python
compare=pd.DataFrame(columns=[df.columns],index=['ROYAL DUTCH SHELL PLC','BP PLC'])
compare
```

```python
compare.loc['ROYAL DUTCH SHELL PLC',df.columns[0:5]]=[21.56,313192,154292,4162,2.12]
compare.loc['BP PLC',df.columns[0:5]]=[4.63,221384,141713,20251,0.35]
compare
```

```python
(compare['Total assets '].values-compare['Total liabilities'].values)
```

```python
compare['Book Value of equity']=(compare['Total assets '].values-compare['Total liabilities'].values)
compare['Book value per share']=(compare['Book Value of equity'].values/compare['Shares outstanding'].values)
compare
```

```python
Valuation=pd.DataFrame(columns=[['Price-to-Book','Price-to-Earning']],index=['Fair ratio','ROYAL DUTCH SHELL PLC','BP PLC'])
Valuation
```

```python
Valuation.loc['Fair ratio','Price-to-Book']=df['Price-to-book(P/B)'].mean()
Valuation.loc['Fair ratio','Price-to-Earning']=df['Price-to-earnings(P/E)'].mean()
Valuation
```

```python
Valuation.loc['Fair ratio','Price-to-Book'].values*compare.loc['ROYAL DUTCH SHELL PLC','Book value per share'].values
```

```python
#Fair Value=fair Price-to-book industry mean*Book value per share
Valuation.loc['ROYAL DUTCH SHELL PLC','Price-to-Book']=Valuation.loc['Fair ratio','Price-to-Book'].values*compare.loc['ROYAL DUTCH SHELL PLC','Book value per share'].values
Valuation.loc['BP PLC','Price-to-Book']=Valuation.loc['Fair ratio','Price-to-Book'].values*compare.loc['BP PLC','Book value per share'].values
Valuation
```

```python
#Fair Value=fair Price-to-Earning industry mean*Earning per share
Valuation.loc['ROYAL DUTCH SHELL PLC','Price-to-Earning']=Valuation.loc['Fair ratio','Price-to-Earning'].values*compare.loc['ROYAL DUTCH SHELL PLC','Earnings per share'].values
Valuation.loc['BP PLC','Price-to-Earning']=Valuation.loc['Fair ratio','Price-to-Earning'].values*compare.loc['BP PLC','Earnings per share'].values
Valuation
```
