```python
#https://simplywall.st/stocks/bd/market-cap-large
import pandas as pd
import numpy as np
import re
```

```python
df.rename(columns = {df.columns[0]:df.columns[-1],df.columns[1]:df.columns[0], df.columns[2]:df.columns[1],
                              df.columns[3]:df.columns[2],df.columns[4]:df.columns[3],df.columns[5]:df.columns[4],
                              df.columns[6]:df.columns[5],df.columns[7]:df.columns[6],
                     df.columns[8]:df.columns[7],df.columns[9]:df.columns[8],
                    df.columns[10]:df.columns[9]}, inplace = True)
df
```

```python
df
```

```python
df=df.drop(df.columns[0],axis=1)
df
```

```python
df.info()
```

```python
for i in df.index:
    x=re.sub(",",'',df.loc[i,'Last Price'])
    df.loc[i,'Last Price']=re.findall("\d+\.\d+",x)[0]
    x2=re.sub(",",'',df.loc[i,'Market Cap'])
    df.loc[i,'Market Cap']=re.findall("\d+\.\d+",x2)[0]
```

```python
df.info()
```

```python
df['Last Price']=df['Last Price'].astype('float')
df['Market Cap']=df['Market Cap'].astype('float')
```
