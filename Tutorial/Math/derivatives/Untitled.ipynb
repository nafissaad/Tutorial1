```python
x, y = np.loadtxt('coviddata.txt')
dydx = np.gradient(y,x)
```

```python
plt.figure(figsize=(10,3), dpi=200)
plt.subplot(1, 2, 1)
plt.plot(x,y,label='$y(x)$')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(x,dydx,label='$y\'(x)$',color='r')
plt.legend()
```

```python
filt=np.ones(15)/15
filt
```

```python
y_smooth=np.convolve(y,filt,mode='valid')
dydx2=np.gradient(y_smooth,x[7:-7])
```

```python
y_smooth
```

```python
plt.figure(figsize=(10,3), dpi=200)
plt.subplot(1, 2, 1)
plt.plot(x,y,label='$y(x)$')
plt.plot(x[7:-7],y_smooth,label=r'$y_{{smooth}}(x)$',color='r')
plt.legend()
plt.subplot(1, 2, 2).set_ylim(-100,120)
plt.plot(x,dydx,label='$y\'(x)$',color='r')
plt.plot(x[7:-7],dydx2,label='$y_{smooth}\'(x)$',color='purple')
plt.legend()
```
