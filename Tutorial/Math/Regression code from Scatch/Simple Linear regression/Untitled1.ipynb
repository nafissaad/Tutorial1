```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
import warnings
```

```python
salary_dataset=pd.read_csv('salary_csv.csv')
salary_dataset
```

```python
x=np.array(salary_dataset['YearsExperience'])
y=np.array(salary_dataset['Salary'])
l=len(x)
```

```python
def cov(a, b):

    if len(a) != len(b):
        return

    a_mean = np.mean(a)
    b_mean = np.mean(b)

    sum = 0

    for i in range(0, len(a)):
        sum += ((a[i] - a_mean) * (b[i] - b_mean))

    return sum/(len(a)-1)
```

```python
cov(x,y)
```
