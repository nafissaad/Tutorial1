```python
import yfinance as yf
import pandas as pd
import datetime as dt
from pandas.tseries.offsets import MonthEnd
```

```python
tickers=pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')[2]
tickers
```

```python
tickers=tickers.Symbol.to_list()
tickers
```

```python
df=yf.download(tickers,start='2009-01-01')
df
```

```python
prices=df['Adj Close']
prices
```

```python
prices.index=pd.to_datetime(prices.index)
```

```python
mlt_ret=prices.pct_change().resample('M').agg(lambda X: (X+1).prod()-1)
mlt_ret
```

```python
formation=dt.datetime(2009,3,31)
```

```python
formation
```

```python
ret_1=mlt_ret.loc[formation-MonthEnd(1)]
```

```python
ret_1
```

```python
formation-MonthEnd(1)
```

```python
ret_1.name='mtl_ret'
```

```python
ret_1
```

```python
ret_1=pd.DataFrame(ret_1)
ret_1
```

```python
ret_1['ranking']=pd.qcut(ret_1.mtl_ret,10,labels=False)
ret_1
```

```python
losers=ret_1[ret_1.ranking==0].index
losers
```

```python
mlt_ret.loc[formation,mlt_ret.columns.isin(losers)]
```

```python
loserret=mlt_ret.loc[formation,mlt_ret.columns.isin(losers)].mean()
loserret
```

```python
def reversal(formation):
    ret_1=mlt_ret.loc[formation-MonthEnd(1)]
    ret_1.name='mtl_ret'
    ret_1=pd.DataFrame(ret_1)
    ret_1['ranking']=pd.qcut(ret_1.mtl_ret,10,labels=False,duplicates='drop')
    losers=ret_1[ret_1.ranking==9].index
    loserret=mlt_ret.loc[formation,mlt_ret.columns.isin(losers)].mean()
    return loserret
```

```python
formation
```

```python
reversal(formation)
```

```python
returns=[]
dates=[]
for i in mlt_ret.index[2:]:
    #print(i)
    returns.append(reversal(i))
    dates.append(i)
```

```python
returns
```

```python
frame=pd.DataFrame({'dates':dates,'Returns':returns})
frame
```

```python
frame.Returns.mean()
```
